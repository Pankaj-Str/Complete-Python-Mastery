```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading the TitanicÂ Dataset
df = sns.load_dataset("titanic")
df.head()
```

```python
# Creating a Pairplot to Explore Relationships

sns.pairplot(df, hue="sex", diag_kind="kde")
plt.show()
```

```python
# Creating a Pairplot to Explore Relationships with different parameters
sns.pairplot(df, hue="pclass", diag_kind="hist")
plt.show()
```

```python
# Visualizing Correlation with aÂ Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

```python
# Distribution of Age with a Histogram
sns.histplot(df['age'].dropna(), kde=True, bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
# Distribution of Age with a Histogram
sns.histplot(df["age"], bins=30, kde=True)
plt.show()
```

```python
# Comparing Passenger Classes with a Box Plot
sns.boxplot(x='pclass', y='age', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Age Distribution Across Passenger Classes')
plt.show()
```

```python
sns.boxplot(x="pclass", y="age", data=df)
plt.show()
```

```python
# Comparing Survival Rates with a ViolinÂ Plot
sns.violinplot(x='survived', y='age', data=df, hue='sex', split=True)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Survival Rates by Age and Gender')
plt.show()
```

```python
# Analyzing Gender and Survival with a Bar Plot
sns.countplot(x='sex', hue='survived', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender and Survival Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# Visualizing Age vs. Fare with a Scatter Plot
sns.scatterplot(x='age', y='fare', data=df, hue='survived', style='sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')
plt.show()
```

```python
# Visualizing Age vs. Fare with a ScatterÂ Plot
sns.scatterplot(x="age", y="fare", hue="pclass", data=df)
plt.show()
```

```python
# Analyzing Trends with a LineÂ Plot
sns.lineplot(x="age", y="survived", hue="pclass", data=df)
plt.show()
```

```python
# Analyzing Trends with a Line Plot
sns.lineplot(x="age", y="fare", hue="pclass", data=df)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare by Passenger Class')
plt.show()
```
