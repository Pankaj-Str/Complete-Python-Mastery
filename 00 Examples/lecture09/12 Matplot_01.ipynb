```python
# using matplot line plot to visualize the data

import matplotlib.pyplot as plt
import numpy as np
# create a student score data
x = np.array([20, 40, 60, 10, 100,45])
y = np.array([2,5,6,8,9,10]) # hours studied
plt.plot(y, x)
plt.xlabel('score')
plt.ylabel('hours studied')
plt.title('Student Score vs Hours Studied')
plt.show()
```

```python
# line plot with different line style and color
plt.plot(x, y, color='green', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=10)
plt.xlabel('score')
plt.ylabel('hours studied')
plt.title('Student Score vs Hours Studied')
plt.show()
# save the plot as a png file
plt.savefig('student_score.png')
```

```python
# Scatter Plot
x = np.array([20, 40, 60, 80, 100])
y = np.array([2,5,6,8,9]) # hours studied
plt.scatter(x, y, color='red', marker='*')
plt.xlabel('score')
plt.ylabel('hours studied')
plt.title('Student Score vs Hours Studied')
plt.show()
```

```python
# create a Bar Plot
x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.array([5,7,3,8,4]) # number of students
plt.bar(x, y, color='blue')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each Grade')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=50, color='purple', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
# add border to the histogram
plt.hist(data, bins=50, color='purple', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
# Pie Chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0,0.1, 0)  # explode
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Grade Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#  Box Plot
data = [np.random.randn(100) + i for i in range(1, 5)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot of Random Data')
plt.show()
```
