```python
# SVM Example with iris

import pandas as pd
from sklearn.datasets import load_iris
```

```python
# load dataset
iris = load_iris()
```

```python
# create a dataframe
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
#check data
df.head()
```

```python
# check lenght of data
print(f'length {len(iris.data)} ')
```

```python
df['target'] = iris.target
print(df.head())
```

```python
# check target 
print(iris.target)
```

```python
# check 1 value
df[df['target'] == 1].head()
```

```python
# check 2 value
df[df['target'] == 2].head()
```

```python
# create flower name column 
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
# check data 
df.head()
```

```python
# set range according to flower name
df1 = df[:50]
df2 = df[50:100]
df3 = df[100:]

print(df1.head())
print(df2.head())
print(df3.head())
```

```python
# visualization
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# sepal lenght / sepal width
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='red',marker='+')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='blue',marker='.')
#plt.scatter(df3['sepal length (cm)'],df2['sepal width (cm)'],color='green',marker='*')
plt.show()
```

```python
# use model 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# set X and y
X = df.drop(['target','flower_name'],axis='columns')
y = df.target
```

```python
# model 
X_train ,X_test ,y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
# check length of x train
print(len(X_train))
# check length of x test
print(len(X_test))
```

```python
# svc model 
model = SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
# model predict
model.predict([[2.8,1.0,1.7,1.3]])
```

```python
# model c value
model_c = SVC(C=7)
model_c.fit(X_train,y_train)
model_c.score(X_test,y_test)
```

```python
# set gannama value
model_c = SVC(gamma=20)
model_c.fit(X_train,y_train)
model_c.score(X_test,y_test)
```
