```python
import seaborn as sns
import matplotlib.pyplot as plt # needed create a plot
import pandas as pd # for use in data table
```

```python
# load dataset iris
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
# create a pairplot 

sns.pairplot(iris,hue="species")
plt.show()
```

```python
sns.pairplot(iris,hue="petal_width")
plt.show()
```

```python
# heatMap
# select only number column 
corr = iris.drop(columns='species').corr()
```

```python
# create a heatmap
sns.heatmap(corr,annot=True,cmap='coolwarm') # annot=True show numbers 
plt.show()
```

```python
# kde plot (kernel density estimate (KDE))
# cerate a KDE plot 
sns.kdeplot(data=iris,x='sepal_length',hue='species') # one line per species 
plt.show()
```

```python
# histogram with KDE

sns.histplot(data=iris,x='sepal_length',kde=True) # kde = True adds a smooth line  
plt.show()
```

```python
# scatter plot 

sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
plt.show()
```

```python
import numpy as np

# line plots 

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# cerate a line plot
sns.lineplot(x=x ,y=y)
plt.show()
```

```python
# box plot

sns.boxplot(data=iris,x='species',y='petal_length')
plt.show()
```
