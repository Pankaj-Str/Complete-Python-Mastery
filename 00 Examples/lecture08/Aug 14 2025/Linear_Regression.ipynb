```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = np.array([8,10,12,25,56,14,78,2,45,36,44,100]).reshape(-1,1) #Independent Variable
y = np.array([10,13,16,15,45,78,96,31,20,14,25,56]) # Dependent Variable
```

```python
# create a train the model
model = LinearRegression()
model.fit(x,y)
```

```python
# get the slope (m) and intercept (b)
m = model.coef_[0] # slope
b = model.intercept_ # intercept
print(f'slope {m}')
print(f'intercept {b}')
```

```python
# predict price for a 20 inch pizza 
x_new = np.array([20]).reshape(-1,1)
y_pred = model.predict(x_new)
print(f" predicted price {y_pred[0]:.2f}")
```

```python
# plot

plt.scatter(x,y, color="blue", label="Data Point")
plt.plot(x,model.predict(x),color="red",label="Regresion Line")
plt.xlabel("Diameter (Inches)")
plt.ylabel("Price : $")
plt.title("pizza price Prediction")
plt.show()

```
