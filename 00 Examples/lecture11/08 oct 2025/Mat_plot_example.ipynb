```python
import matplotlib.pyplot as plt

sales = [1200,3400,4500,450,670,2300,6700,230,10000]
employee_id = [101,102,103,104,105,106,107,108,109]

plt.plot(employee_id,sales)
plt.title("Sales Data")
plt.xlabel("Sales")
plt.ylabel("Employee ID")
plt.show()
```

```python
plt.scatter(employee_id,sales)
plt.title("Sales Data")
plt.xlabel("Sales")
plt.ylabel("Employee ID")
plt.show()
```

```python
plt.bar(employee_id,sales)
plt.title("Sales Data")
plt.xlabel("Sales")
plt.ylabel("Employee ID")
plt.show()
```

```python
plt.hist(sales,bins=10,color='skyblue',edgecolor='black')
plt.title("Sales Data")
plt.xlabel("Sales")
plt.ylabel("Employee ID")
plt.show()
```

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','blue','green','red','pink','yellow']
plt.pie(sales, labels=employee_id , colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie
plt.title("Sales Data")
plt.show()
```

```python
plt.boxplot(sales, vert=True, patch_artist=True)
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
