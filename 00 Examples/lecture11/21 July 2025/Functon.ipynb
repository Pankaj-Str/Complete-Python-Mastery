```python
def bill(price,tax):
    gst_amount = price*tax/100
    final_price = gst_amount + price
    print("Price = ₹",price,"/-")
    print("Tax = ",tax,"%")
    print("GST Amount = ₹",gst_amount,"/-")
    print("MRP = ₹",final_price,"/-")
    print("-------------------------")
```

```python
bill(4500,18)
bill(1200,2)
bill(7000,9)
```
