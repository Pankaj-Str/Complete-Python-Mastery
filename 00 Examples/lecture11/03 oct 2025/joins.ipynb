```python
import pandas as pd

# load dataset

data1 = pd.read_csv('credit_card_transactions.csv')
data2 = pd.read_csv('customer_info.csv')
```

```python
data1.head()
```

```python
data2.head()
```

```python
inner_join = pd.merge(data1,data2, how='inner')

inner_join.to_csv('inner_join.csv')
```

```python
left_join = pd.merge(data1,data2, how='left')

left_join.to_csv('left_join.csv')
```

```python
right_join = pd.merge(data1,data2, how='right')

right_join.to_csv('right_join.csv')
```

```python
outer_join = pd.merge(data1,data2, how='outer')

outer_join.to_csv('outer_join.csv')
```
