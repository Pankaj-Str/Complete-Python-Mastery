```python
# check version
import pandas as pd
print(pd.__version__)
```

```python
# Series Creation
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
```

```python
# change index of series
custom_index = ['a', 'b', 'c', 'd', 'e']
data = pd.Series(data, index=custom_index)
print(data)
```

```python
# using dictionary to create series
data_dict = {'name':'Nishant', 'age':25, 'city':'New York'}
data_s = pd.Series(data_dict)
print(data_s)
```

```python
# accessing elements in series
print(data_s['name'])

# accessing multiple elements
print(data_s[['name', 'city']])
```

```python
# DataFrame Creation

data = {
    'Name':['Nishant','Joy','Ashil','Anil','tiya','Jiya'],
    'Age':[25,24,23,22,21,20],
    'City':['New York','Los Angeles','Chicago','Houston','Phoenix','Philadelphia'],
    'Course':['Python','Java','C++','JavaScript','Ruby','Go']
}
```

```python
# create DataFrame

df = pd.DataFrame(data)
print(df)
```

```python
# change column name 
df.columns = ['Student Name', 'Student Age', 'Student City', 'Programming Course']
print(df)
```

```python
# change index in dataframe
df.index = ['s1', 's2', 's3', 's4', 's5', 's6']
print(df)
```

```python
# https://cwpc.in/mastering-pandas-dataframes-a-comprehensive-guide-codes-with-pankaj-63b22098a6d4
```

```python
# add new row in dataframe using concat
new_row = pd.DataFrame({'Student Name':['Riya'],
                        'Student Age':[19],
                        'Student City':['San Antonio'],
                        'Programming Course':['Swift']}, index=['s7'])
df = pd.concat([df, new_row])
print(df)
```

```python
# add new column to dataframe
df['Graduation Year'] = [2022, 2023, 2024, 2025, 2026, 2027, 2028]
print(df)   
```

```python
# export dataframe to csv
df.to_csv('students_data.csv', index=False)
```
