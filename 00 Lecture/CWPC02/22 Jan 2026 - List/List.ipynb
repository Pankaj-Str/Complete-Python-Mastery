```python
data = [12,23,24,25,56]
# access data from list 
print(data[2])
```

```python
# Slicing data 
print(data[2:4])
print(data[:4])
print(data[2:])
```

```python
# change elements 
data[1] = 99
print(data)
```

```python
#append()
data.append(800)
print(data)
```

```python
# insert add at specific index 
data.insert(1,900)
print(data)
```

```python
# extend data - add multiple elements 
data.extend([5,6,7])
print(data)
```

```python
# find lenght
print(len(data))
```

```python
# total 
sum(data)
```

```python
# find min 
min(data)
```

```python
# find max 
max(data)
```

```python
# print using loop 
for i in data:
    print(i)
```
