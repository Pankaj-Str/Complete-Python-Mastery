```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# data set 
data = {
    'Size':[1000, 1500, 2000, 2500, 3000], # house size in square feet
    'Bedrooms':[2, 3, 4, 3, 5], # number of bedrooms
    'Price':[200000, 250000, 300000, 350000, 400000] # price of the house
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# independent variables (features)
X = df[['Size', 'Bedrooms']]
```

```python
# dependent variable (target)
y = df['Price']
```

```python
# create and fit the model
model = LinearRegression()
model.fit(X, y)
```

```python
# check  predictions
predictions = model.predict([[1600, 3]]) # predict the price of a house with 1600 sqft and 4 bedrooms
print(predictions)
```

```python
# use live dataset 
# load dataset

# url = 'https://raw.githubusercontent.com/Pankaj-Str/MachineLearning/refs/heads/main/DataSet/house-prices-advanced-regression-techniques/train.csv'

df = pd.read_csv("train.csv")
df.head()
```

```python
X = df[['LotArea', 'OverallQual', 'YearBuilt']]
y = df['SalePrice']
```

```python
# model 
model = LinearRegression()
model.fit(X, y)
```

```python
# predictions
predictions = model.predict([[8450, 7, 2003]]) # predict the price of a house with 8450 sqft lot area, overall quality of 7, and built in 2003
print(predictions)
```

```python
# visualization
import matplotlib.pyplot as plt
plt.scatter(df['LotArea'], df['SalePrice'], color='blue', label='Actual')
plt.scatter(df['LotArea'], model.predict(X), color='red', label='Predicted')
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.title('Actual vs Predicted Sale Price')
plt.legend()
plt.show()
```
