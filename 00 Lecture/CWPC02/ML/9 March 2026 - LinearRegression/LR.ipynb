```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = np.array([8,10,12]).reshape(-1, 1) # pizza diameter
y = np.array([10,13,16]) # pizza price
```

```python
# create and fit the model
model = LinearRegression()
model.fit(X, y)
```

```python
# get the  slope (m) and intercept (b)
m = model.coef_[0]
b = model.intercept_
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```

```python
# predict the price of a pizza with diameter 20 inches
X_new = np.array([[20]])
y_pred = model.predict(X_new)
print(f"Predicted price for a 20 inch pizza: {y_pred[0]}")
```

```python
plt.scatter(X , y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Pizza Diameter (inches)')
plt.ylabel('Pizza Price ($)')
plt.title('Linear Regression: Pizza Price vs Diameter')
plt.legend()
plt.show()
    
```
