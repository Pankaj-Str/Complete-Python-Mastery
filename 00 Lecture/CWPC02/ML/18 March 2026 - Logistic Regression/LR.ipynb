```python
import math

# sigmoid function 
def sigmoid(z):
    return 1 / (1 + math.exp(-z))
```

```python
hours = 5
z = -1.5 + 0.6 * hours
chance = sigmoid(z)
prediction = "pass" if chance  > 0.5 else "fail" # decide
print(f"chance of passing {chance:.2f} that's {chance*100:.0f}%")
print(f"Prediction = {prediction}")
```
