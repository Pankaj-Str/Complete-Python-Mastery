```python
# convert Transaction_Date to date format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'],errors='coerce')
```

```python
# check datatype
df.dtypes
```

```python
# add year , month , and day 
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day
```

```python
df.head()
```

```python
# check info
df.info()
```

```python
# find the Transactions january 2023 
jan_23 = df[(df['Year'] == 2023)& (df['Month']==1)]
print(jan_23.head())
```

```python
# find the transactions with amount > 2500 transaction type = online 
online_1000 = df[(df['Amount'] > 2500) & (df['Transaction_Type']=='Online')]
# print(online_1000)

# create csv 
online_1000.to_csv('online2500.csv')
```
