```python
from sklearn import svm
import numpy as np
```

```python
# data X features and y labels
X = np.array([[0, 0], [1, 1]]) # 2 data points with 2 features each
y = np.array([0, 1]) # corresponding labels for the data points
# 0 = class 0, 1 = class 1
```

```python
# create SVM model
model = svm.SVC(kernel='linear') # using linear kernel
model.fit(X, y) # train the model on the data
```

```python
# predicting new data points
new_data = np.array([[0.5, 0.5], [1.5, 1.5]]) # new data points to classify
predictions = model.predict(new_data) # predict the class labels for the new data points
print(predictions) # output the predicted class labels for the new data points
```

```python
# Example - 2nd 
import numpy as np
import matplotlib.pyplot as plt
# dataset 
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# load the iris dataset
iris = load_iris()
X = iris.data # features
y = iris.target # labels
```

```python
# split the dataset into training and testing sets
# 30%  test data, 70% training data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# create SVM model
model = svm.SVC(kernel='linear', C=1) # using linear kernel
model.fit(X_train, y_train) # train the model on the training data
```

```python
# make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# check accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis') # plot the data points colored by their class labels
plt.title('Iris Dataset - SVM Classification')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
