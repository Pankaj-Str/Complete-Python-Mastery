```python
import numpy as np
```

```python
# most common activation functions are sigmoid, tanh, and relu.
def relu(x):
    return np.maximum(0, x) # if x > 0 keep it , else make it 0

def sigmoid(x):
    return 1 / (1 + np.exp(-x)) # if x is large then it will be 1, if x is small then it will be 0

def linear(x):
    return x # linear activation function is used for regression problems
```

```python
# fake neural example 
input = np.array([1.0,-2.0,0.5]) # data input to the neural network
weights = np.array([0.5,0.2,0.1]) # learning weights of the neural network
bias = 0.1 # bias of the neural network
```

```python
# step 1 weighted sum of inputs
z = np.dot(input, weights) + bias # weighted sum of inputs

print("row calculation of weighted sum of inputs: ", z)
```

```python
print("linear activation function output: ", linear(z))
print("relu activation function output: ", relu(z))
print("sigmoid activation function output: ", sigmoid(z))
print("sigmoid rounded activation function output: ", round(sigmoid(z), 2))
```

```python
# layer 1

w1 = np.random.rand(3, 4)* 0.1 # weights for layer 1
b1 = np.random.rand(4)* 0.1 # bias for layer 1

# layer 2
w2 = np.random.rand(4, 1)* 0.1 # weights for layer 2
b2 = np.random.rand(1)* 0.1 # bias for layer 2


# forward pass
z1 = np.dot(input, w1) + b1 # weighted sum of inputs for layer 1
a1 = relu(z1) # activation function for layer 1

print("layer 1 output: ", a1)

# sigmoid activation function for layer 1
a1_sigmoid = sigmoid(z1)
print("layer 1 output with sigmoid activation function: ", a1_sigmoid)
```
