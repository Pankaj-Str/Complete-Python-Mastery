```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')
print(data.head()) # top 5 rows of the dataset
```

```python
# check column names and data types
print(data.info())
```

```python
# check column names
data.columns
```

```python
# rename columns for easier access
data = data.rename(
    columns={
        ' Open': 'Open', 
        ' High': 'High', 
        ' Low': 'Low', 
        ' Close': 'Close', 
})
```

```python
# check column name again
data.columns
```

```python
# convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# create a line plot of 'Close' price over time
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
# create a line plot of 'Open' price over time
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Open'], label='Close Price')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Open'], label='Close Price')
plt.plot(data['Date'], data['Close'], label='Close Price')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
# bar plot of 'Close' price by month
data['Month'] = data['Date'].dt.month # find the month from the 'Date' column
monthly_close = data.groupby('Month')['Close'].mean()
plt.figure(figsize=(10, 5))
monthly_close.plot(kind='bar')
plt.title('Average Close Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.grid()
plt.show()
```
