```python
Employee = {
    "Name":"Joy",
    "Age":22,
    "City":"Mumbai"
}
```

```python
# print data from Dictionary
print(Employee)
```

```python
# Accessing Values 
print(Employee['Name'])
```

```python
print(Employee.get("Age"))
```

```python
# add new item into list - 
Employee['Email'] = "joy@codeswithpankaj.com"
```

```python
print(Employee)
```

```python
# update 
Employee['Age'] = 45
print(Employee)
```

```python
#removing items 
Employee.pop("Age")
print(Employee)
```

```python
del Employee['City']
print(Employee)
```

```python
# find all keys 
Employee.keys()
```

```python
# find all values 
Employee.values()
```

```python
# access all items 
Employee.items()
```

```python
for key,value in Employee.items():
    print(key,' = ',value)
```

```python
# find lenght
len(Employee)
```
