```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock_symbol = 'RELIANCE.NS'
start_date = '2024-01-01'
end_date = '2025-11-17'

data = yf.download(stock_symbol,start_date,end_date)

data.to_csv("stock.csv")
```

```python
data.head() 
```

```python
# view summary 
data.describe()
```

```python
print(data.isnull().sum())
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['High'], label='Closing Price',color='blue')
plt.plot(data['Low'], label='Closing Price',color='Green')
plt.title("RELIANCE.NS Closing Price ")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()
```

```python
data.tail()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Closing Price',color='blue')
plt.plot(data['MA50'], label='Closing Price',color='Green')
plt.plot(data['MA200'], label='Closing Price',color='pink')
plt.title("RELIANCE.NS Closing Price ")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.show()
```
