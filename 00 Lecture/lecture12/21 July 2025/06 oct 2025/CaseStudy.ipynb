```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# load data 
data = pd.read_csv("HistoricalPrices.csv")
print(data.head())
```

```python
data.info()
```

```python
data.columns
```

```python
# clean data 
data = data.rename(columns={' Open':'Open', ' High':'High', ' Low':'Low', ' Close':'Close'})
```

```python
data.columns
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values(by='Date')
```

```python
# line plot
plt.plot(data['Date'],data['Close'])
plt.show()
```

```python
plt.plot(data['Date'],data['Close'])
plt.plot(data['Date'],data['Open'])
plt.plot(data['Date'],data['High'])
plt.plot(data['Date'],data['Low'])
plt.show()
```

```python
plt.plot(data['Date'],data['Close'] , label = 'Open')
plt.plot(data['Date'],data['Open'] , label= 'Close')
plt.legend()
plt.show()

# bar plots
plt.bar(data['Date'],data['Close'])
plt.show()

# plots month wise 
data['Month'] = data['Date'].dt.month
monthly_averages = data.groupby('Month').mean()
plt.bar(monthly_averages.index,monthly_averages['Close'])
plt.show()

from calendar import month_name
data['Month'] = pd.Categorical(data['Date'].dt.month_name(),month_name[:])
data_monthly_mean = data.groupby('Month').mean().reset_index()
data_monthly_mean.head()

plt.bar(data_monthly_mean['Month'],height=data_monthly_mean['Close'])
plt.show()
# horizontal Bar Plots
plt.barh(data_monthly_mean['Month'],width=data_monthly_mean['Close'])
plt.show()
```
