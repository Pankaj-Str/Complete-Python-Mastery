```python
import matplotlib.pyplot as plt
import numpy as np
sales = np.array([10,20,30,40,45,50,60,65])
price = np.array([1000,4000,3000,500,300,200,100,5000])
plt.plot(sales,price)
plt.title("Sales Data")
plt.xlabel("Sales")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
categories = ['A','B','C','D','E','F','G','H']
Sales = np.array([1000,4000,3000,500,300,200,100,5000])
plt.bar(categories,Sales)
plt.title("categories Sales Data")
plt.xlabel("Sales")
plt.ylabel("categories")
plt.grid(True)
plt.show()
```

```python
sales = np.array([10,20,30,40,45,50,60,65])
price = np.array([1000,4000,3000,500,300,200,100,5000])

size = np.array([100,200,300,400,450,500,600,650])

colors = ['red','blue','green','orange','purple','black','pink','yellow']

plt.scatter(sales,price, color=colors,s=size,alpha=0.8)
plt.title("categories Sales Data")
plt.xlabel("Sales")
plt.ylabel("categories")
plt.grid(True)
plt.show()
```

```python

#price = np.array([1000,4000,3000,500,300,200,100,5000])
np.random.seed(42) # fix data
score = np.random.normal(75,10,200)
plt.hist(score,bins=30,color='skyblue',edgecolor='black')
plt.title("categories Sales Data")
plt.xlabel("Sales")
plt.ylabel("categories")
plt.grid(axis='y',alpha=0.5,linestyle='--')
plt.show()
```
