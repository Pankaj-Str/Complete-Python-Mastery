```python
data = {21,21,23,23,44,56,67,56,56}
print(data)
```

```python
# using set() method 
my_list_data = [34,34,34,45,56,67,78,89,90]
data = set(my_list_data)
print(data)
```

```python
# using add() method 
data.add(100)
print(data)
```

```python
# update() method
data.update([200,300,400])
print(data)
```

```python
# remove() method
data.remove(200)
```

```python
print(data)
```

```python
# discard() method
data.discard(45)
print(data)
```

```python
# pop() method
data.pop()
print(data)
```

```python
# clear() method
data.clear()
print(data)
```

```python
# Checking Membership
data = {21,21,23,23,44,56,67,56,56}
print(21 in data)
print(100 in data)
```

```python
# Iterating (Looping)
for i in data:
    print(i)
```

```python
# copy() method
data_copy = data.copy()
print(data_copy)
```

```python
# len() method
print(len(data))
```
