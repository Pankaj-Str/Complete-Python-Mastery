```python
# create a list 

fruits = ["apple", "banana", "cherry"]
print(fruits)
```

```python
# # accessing list items
print(fruits[1])
print(fruits[-1])
```

```python
# update list items
fruits[0] = "orange"
print(fruits)
```

```python
# slice list
print(fruits[1:3])
print(fruits[:2])
print(fruits[1:])
print(fruits[:])
```

```python
# List Methods
# append() -> add an item to the end of the list
fruits.append("grape")
print(fruits)
```

```python
# extend() -> add all items of a list to another list
more_fruits = ["kiwi", "melon"]
fruits.extend(more_fruits)
print(fruits)
```

```python
# insert() -> add an item at a specified index
fruits.insert(1, "strawberry")
print(fruits)
```

```python
# remove() -> remove the first occurrence of an item
fruits.remove("banana")
print(fruits)
```

```python
# pop() -> remove an item at a specified index (default is the last item)
fruits.pop(2)
print(fruits)
```

```python
fruits.pop()
print(fruits)
```

```python
# index() -> return the index of the first occurrence of an item
print(fruits.index("kiwi"))

```

```python
# count() -> return the number of occurrences of an item
print(fruits.count("kiwi"))
```

```python
data = [1, 2, 3, 4, 5,4,2,1,2,3,4,5]
# sort() -> sort the items of the list in ascending order
data.sort()
print(data)

```

```python
# reverse() -> reverse the order of the items in the list
data.reverse()
print(data)
```

```python
# copy() -> return a copy of the list
d = data.copy()
print(d)
```

```python
# clear() -> remove all items from the list
data.clear()
print(data)
```

```python
# using loop to print list items
for i in fruits:
    print(i)
```
