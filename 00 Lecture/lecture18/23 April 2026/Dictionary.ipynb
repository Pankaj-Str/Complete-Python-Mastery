```python
# Dictionary 

# create a dictionary
data = {} # empty dictionary
print(data) 

# check the type of data
print(type(data))
```

```python
# create a info dictionary

info = {
    'name':'joy',
    'age': 25,
    'city': 'Mumbai'
}

print(info)
```

```python
# accessing values from dictionary
print(info['name'])
print(info['age']) 
```

```python
# modifying values in dictionary
info['age'] = 26
print(info)
```

```python
# adding new key-value pair to dictionary
info['height'] = 5.8
print(info)
```

```python
# using loop to access dictionary

for i in info:
    print(i,' = ',info[i])
```

```python
# keys(): Returns a list of all keys in the dictionary.
print(info.keys())
# values(): Returns a list of all values in the dictionary.
print(info.values())
# items(): Returns a list of key-value pairs as tuples.
print(info.items())
```

```python
# get(key, default): Returns the value for a key if it exists, else returns a default value.
print(info.get('name'))
print(info.get('country', 'Not Found'))
```

```python
# pop(key, default): Removes and returns the value associated with the key.
print(info.pop('city'))
print(info)
```
