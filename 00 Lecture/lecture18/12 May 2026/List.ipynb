```python
# -- User input section
# Enter Product List Size : 4
# Enter Product 1 -
# Samosa
# Enter Product 2 -
# Kachori
# Enter Product 3 -
# Fafda
# Enter Product 4 -
# Jalebi

# Enter Samosa Price :
# 300/-
# Enter Kachori Price :
# 100/-
# Enter Fafda Price :
# 100/-
# Enter Jalebi Price :
# 200/-

# --- output

# samosa = 300
# kachori = 100
# fafda = 100
# jalebi = 200
```

```python
size = int(input("Enter Product List Size : "))
product_list = []

for i in range(size):
    product = input(f"Enter Product {i+1} -\n")
    product_list.append(product)
    
price_list = []
for i in range(size):
    price = input(f"Enter {product_list[i]} Price :\n")
    price_list.append(price)
    
    
for i in range(size):
    print(f"{product_list[i]} = {price_list[i]}")     
```
