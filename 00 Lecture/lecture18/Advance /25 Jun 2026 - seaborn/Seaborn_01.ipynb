```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# set style of seaborn
sns.set_theme(style="whitegrid",palette="tab10")
# ['white', 'dark', 'whitegrid', 'darkgrid', 'ticks']
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
print(tips.info())
```

```python
tips.describe()
```

```python
# scatter plot 
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip")
plt.title("total bill vs tips")
plt.show()
```

```python
# scatter plot 
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip",
                hue="sex",
                style="smoker",
                size="size",
                palette="deep")
plt.title("total bill vs tips")
plt.show()
```

```python
# line plot 
flights = sns.load_dataset("Flights")
flights.head()
```

```python
# line plot 
sns.lineplot(data=flights,
             x='year',
             y='passengers',
             hue='month')
plt.title("Airline passengers over year ")
plt.show()
```

```python
# histogram + KDE
sns.histplot(data=tips,x="total_bill",bins=30,kde=True)
plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill",hue="sex",estimator=np.mean)
plt.show()
```

```python
sns.countplot(data=tips,x="day",hue="sex")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex")
plt.show()
```
