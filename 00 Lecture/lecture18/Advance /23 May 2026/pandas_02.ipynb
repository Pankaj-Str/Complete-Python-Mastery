```python
# read csv file 
# step 1 import pandas library
import pandas as pd
```

```python
# step 2 read csv file using pandas library

df = pd.read_csv('sales.csv')
```

```python
# step 3 - print only 10 rows of the dataframe
df.head(10)  # head() default 5 rows print karta hai, agar hum 10 denge to 10 rows print karega.
```

```python
# step 4 check data size 
# number of rows and columns
print(df.shape)
```

```python
# step 5 - check column names
print(df.columns.tolist())

for i in df.columns:
    print(i)
```

```python
# check data types of each column
print(df.dtypes)
```

```python
# check all info about the dataframe
print(df.info())
```

```python
# check missing values in the dataframe
print(df.isnull().sum())
```

```python
# check statistics of the dataframe
print(df.describe())
```

```python
# print only 3 columns of the dataframe
df2 = df[['Product', 'Unit_Price', 'Quantity']]
print(df2.head()) # top 5 rows print karega by default

print("----------------------")
print(df2.tail()) # last 5 rows print karega by default
```

```python
# find the minimum unit price of the products
print(df['Unit_Price'].min())

# find the maximum unit price of the products
print(df['Unit_Price'].max())
```

```python
# find the average unit price of the products
print(df['Unit_Price'].mean())
```

```python
# find unit_price is greater than 2000
new_df = df[df['Unit_Price'] > 2000]
print(new_df)

# save reported data in new csv file
new_df.to_csv('expensive_products.csv', index=False)
```

```python
# group by 

# total sales city wise
city_sales = df.groupby('City')['Total_Amount'].sum()
print(city_sales)

# total avg city wise
city_sales_avg = df.groupby('City')['Total_Amount'].mean()
print(city_sales_avg)
```

```python
# top 2 cities with highest sales
top_cities = city_sales_avg.sort_values(ascending=False).head(2)
print(top_cities)
```

```python
# multiple statistics city wise

result = df.groupby('City').agg({
    'Total_Amount': ['sum', 'mean', 'min', 'max'],
    'Quantity': ['sum', 'mean']
}).round(2)

print(result)
```

```python
# group two columns city and product wise

result2 = df.groupby(['City', 'Category'])['Unit_Price'].sum().reset_index()
print(result2.head())

# find top 5 products with highest unit price city wise
result3 = result2.sort_values(by='Unit_Price', ascending=False).head(5)
print(result3)
```
