```python
import pandas as pd

print(pd.__version__)
```

```python
data = ['apple', 'banana', 'cherry','watermelon','grapes']

# series()

s = pd.Series(data)
print(s)
```

```python
# update index 
x = ['a','b','c','d','e']

s = pd.Series(data,index=x)
print(s)
```

```python
# accessing data using index
print(s[0])

print(s['a'])
```

```python
# dataframe()

data = {
    'name':['joy','eha','shubham','vivek','vedant'],
    'age':[23,24,25,26,27],
    'city':['delhi','mumbai','pune','bangalore','chennai']
}

df = pd.DataFrame(data)
df
```

```python
# update the index
x = ['a','b','c','d','e']
df = pd.DataFrame(data,index=x)
df
```
