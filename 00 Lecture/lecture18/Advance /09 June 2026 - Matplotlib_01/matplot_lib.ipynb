```python
import matplotlib.pyplot as plt
```

```python
# line plot 

sales = [100, 170, 450, 150, 350]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

plt.plot(months, sales , color="red" , linestyle='--',marker='o')
plt.title("sales by month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# bar chart 
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 170, 450, 150, 350]
plt.bar(months,sales,color=['blue','pink','green','yellow','lightgreen'])
plt.title("sales by month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
sales = [100, 170, 450, 150, 350]
values = [2,3,4,6,7]

plt.scatter(sales,values)
plt.show()
```

```python
import numpy as np
np.random.seed(42)
x = np.random.rand(100)

y = 2 * x + np.random.randn(100) * 0.3

plt.figure(figsize=(8,5))
plt.scatter(x,y,c=np.random.rand(100),s=100*np.random.rand(100),alpha=0.6,cmap='plasma')
plt.grid(True,alpha=0.5)
plt.colorbar(label='Color Codes')
plt.show()
```

```python
# pie chart

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [10, 17, 45, 15, 35]
explode = [0.3, 0, 0, 0, 0]  # explode the first slice

plt.figure(figsize=(6,6))
plt.pie(sales, labels=months,explode= explode , autopct='%1.2f%%',shadow=True,startangle=180,
        colors=['lightblue', 'lightcoral', 'lightgreen', 'lightyellow', 'lightpink'])
plt.title("Sales Distribution by Month")
plt.show()
```
