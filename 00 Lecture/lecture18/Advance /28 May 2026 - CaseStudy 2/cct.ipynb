#### 1️⃣ Data Exploration and Cleaning
- Load the dataset into a Pandas DataFrame and display the first 5 rows.
- Check the shape, column names, and summary statistics of the dataset.
- Identify and handle missing values (fill or drop based on the data type).
- Convert Transaction_Date into datetime format and extract year, month, and day as new columns.

```python
df = pd.read_csv('clean_cc.csv')
# print first 5 rows of the dataframe
df.head()
```

```python
# Check the shape, column names, and summary statistics of the dataset.
print(df.shape)
print(df.columns)
print(df.describe())
```

```python
# Identify and handle missing values (fill or drop based on the data type).
print(df.dtypes)
```

```python
# check missing values in each column
print(df.isnull().sum())
```

```python
# delete location missing values rows 
df = df.dropna(subset=['Location'])
# check again missing values in each column
print(df.isnull().sum())
```

```python
print('check size of rows ',df.shape[0])
```

```python
# Handle missing
# fill mean value into Amount column
mean_amount = df['Amount'].mean()
df['Amount'] = df['Amount'].fillna(mean_amount)
# check again missing values in each column
print(df.isnull().sum())
```

```python
# convert date column to datetime format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
# check the data types again
print(df.dtypes)
```

```python
# handle missing value 
df['Transaction_Type'] = df['Transaction_Type'].fillna('Unknown')
df['Payment_Mode'] = df['Payment_Mode'].fillna('Unknown')
df['Transaction_Status'] = df['Transaction_Status'].fillna('Unknown')
# check again missing values in each column
print(df.isnull().sum())
```

```python
# Extract year, month, and day as new columns.
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day

df.head()
```

```python
# Retrieve all transactions made in January 2023.
jan_2023 = df[(df['Year'] == 2023) & (df['Month'] == 1)]
# print(jan_2023.head())
print("total ",jan_2023.shape[0])
jan_2023.to_csv("jan_2023.csv")
```
