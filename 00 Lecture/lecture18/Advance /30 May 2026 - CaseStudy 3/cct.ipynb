```python
df = pd.read_csv('clean_cc.csv')
# print first 5 rows of the dataframe
df.head()
```

```python
# Check the shape, column names, and summary statistics of the dataset.
print(df.shape)
print(df.columns)
print(df.describe())
```

```python
# Identify and handle missing values (fill or drop based on the data type).
print(df.dtypes)
```

```python
# check missing values in each column
print(df.isnull().sum())
```

```python
# delete location missing values rows 
df = df.dropna(subset=['Location'])
# check again missing values in each column
print(df.isnull().sum())
```

```python
print('check size of rows ',df.shape[0])
```

```python
# Handle missing
# fill mean value into Amount column
mean_amount = df['Amount'].mean()
df['Amount'] = df['Amount'].fillna(mean_amount)
# check again missing values in each column
print(df.isnull().sum())
```

```python
# convert date column to datetime format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
# check the data types again
print(df.dtypes)
```

```python
# handle missing value 
df['Transaction_Type'] = df['Transaction_Type'].fillna('Unknown')
df['Payment_Mode'] = df['Payment_Mode'].fillna('Unknown')
df['Transaction_Status'] = df['Transaction_Status'].fillna('Unknown')
# check again missing values in each column
print(df.isnull().sum())
```

```python
# Extract year, month, and day as new columns.
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day

df.head()
```

```python
# Retrieve all transactions made in January 2023.
jan_2023 = df[(df['Year'] == 2023) & (df['Month'] == 1)]
# print(jan_2023.head())
print("total ",jan_2023.shape[0])
jan_2023.to_csv("jan_2023.csv")
```

```python
# Create a new column Discounted_Amount, assuming a 5% discount on all transactions above $2000.
# apply() -> new column 
df['Discounted_Amount'] = df['Amount'].apply(lambda x: x * 0.95 if x > 2000 else x)

print(df[['Amount','Discounted_Amount']].head())
```

```python
# Categorize the Transaction_Amount into Low, Medium, and High based on:
    # Low: Below $2000
    # Medium: Between $2000 - $3500
    # High: Above $3500
    
#cut() -> Categorize

df['Amount_Category'] = pd.cut(df['Amount'],
                               bins=[-float('inf'),2000,3500,float('inf')],
                               labels=['Low','Medium','High'])

print(df[['Amount','Amount_Category']].head())
print(df[['Amount','Amount_Category']].tail())
```

```python
# Drop the Merchant column if more than 30% of values are missing.
missing_Merchant_percentage = df['Merchant'].isnull().mean()
print(missing_Merchant_percentage)

if missing_Merchant_percentage > 0.3:
    df.drop(columns=['Merchant'])
else:
    print("Done ....")    
```
