```python
# import libraries
import seaborn as sns

# check seaborn version
print(sns.__version__)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load dataset
df = sns.load_dataset('iris')
df.head()
```

```python
# create a pairplot
sns.pairplot(df, hue='species')
plt.show()
```

```python
# create a heatmap
# calculate correlation matrix (ONLY numerical columns)
corr = df.drop('species', axis=1).corr()
corr.head()
```

```python
# now create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# pairplot for two columns
sns.pairplot(df, hue='species', vars=['sepal_length', 'sepal_width'])
plt.show()
```

```python
# kernel density estimate plot (KDE plot)
sns.kdeplot(data=df, x='sepal_length', hue='species', fill=True)
plt.show()
```

```python
# visualizing Distributions with sns.histplot
sns.histplot(data=df, x='sepal_length', hue='species', kde=True, element='step', stat='density', common_norm=False)
plt.show()
```

```python
# scatter plot
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', style='species', s=100)
plt.show()
```

```python
# line plot
sns.lineplot(data=df, x='sepal_length', y='sepal_width', hue='species', marker='o')
plt.show()
```

```python
# histogram with multiple categories
sns.histplot(data=df, x='sepal_length', hue='species', multiple='stack', bins=20)
plt.show()
```

```python
# box plot
sns.boxplot(data=df, x='species', y='sepal_length')
plt.show()
```

```python
# violin plot
sns.violinplot(data=df, x='species', y='sepal_length')
plt.show()
```
