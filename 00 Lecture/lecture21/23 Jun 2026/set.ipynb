```python
# create a empty set 
data = set()
print(data)
# check type 
print(type(data))
```

```python
# create a set with some values

numbers = {12,13,11,11,113,13,1,3,13,13,13,111,113,14,14,14}
print(numbers)
```

```python
# create String set

languages = {"Python", "Java", "C++", "C#", "JavaScript", "Python","Python","Java"}

print(languages)
```

```python
# add new element to set
languages.add("Go")
print(languages)
```

```python
# update set with multiple values
languages.update(["Ruby", "Rust", "Kotlin"])
print(languages)
```

```python
# delete an element from set
languages.remove("Java")
print(languages)
```

```python
# discard an element from set
languages.discard("C++")
print(languages)
```

```python
# pop an element from set
languages.pop()
print(languages)
```

```python
# clear all elements from set
languages.clear()
print(languages)
```

```python
# using in keyword to check if an element is present in set

data = {'Kotlin', 'Python', 'Go', 'Ruby', 'JavaScript', 'C#'}

if 'Python' in data:
    print("Python is present in set")
else :
    print("Python is not present in set")
```

```python
# using loop 
for i in data:
    print(i)
```

```python
# find the set size 
print(len(data))
```
