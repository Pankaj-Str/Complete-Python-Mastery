```python
# create a dictionary
# info = {'name': 'Nishant', 'age': 21, 'height': 4.5}
info = {
    'name':'Nishant',
    'age':21,
    'height':4.5    
}
print(info)
```

```python
# access data
print(info['name'])
```

```python
# update data 
info['age'] = 34
print(info)
```

```python
# add new item
info['city'] = 'mumbai'
print(info)
```

```python
# Checking Key Existence
if "age" in info:
    print("age is find...")
```

```python
# keys(): Returns a list of all keys in the dictionary.
print(info.keys())
# values(): Returns a list of all values in the dictionary.
print(info.values())
# items(): Returns a list of key-value pairs as tuples.
print(info.items())
# get(key, default): Returns the value for a key if it exists, else returns a default value.
print(info.get('age'))
```

```python
# pop(key, default): Removes and returns the value associated with the key.
info.pop('name')
print(info)
```

```python
# using loop to print 
for i in info:
    print(f"{i} = {info[i]}")
```
