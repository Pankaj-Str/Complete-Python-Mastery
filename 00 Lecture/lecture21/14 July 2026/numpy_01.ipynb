```python
import numpy as np

print(np.__version__)
```

```python
# create a single array 1D 

arr_1 = np.array([23,24,25,26,27,28,29])
print(arr_1)

# check type 
print(type(arr_1))
# dimensions
print(np.ndim(arr_1))
# check size 
print(arr_1.size)
# check column and row
print(arr_1.shape)

# access data from array 

print(arr_1[3])
```

```python
arr_1 = np.array([[12,13,14],[34,35,36],[45,46,47]])
print(arr_1)

# check type 
print(type(arr_1))
# dimensions
print(np.ndim(arr_1))
# check size 
print(arr_1.size)
# check column and row
print(arr_1.shape) # (row,col)

print(arr_1[1,1])
```

```python
arr_1 = np.array([[
                   [12,13,14],
                   [34,35,36],
                   [45,46,47]],
                  
                  [[52,53,54],
                   [24,55,56],
                   [55,56,57]]])

print(arr_1)

# check type 
print(type(arr_1))
# dimensions
print(np.ndim(arr_1))
# check size 
print(arr_1.size)
# check column and row
print(arr_1.shape) # (no_of_array,row,col)

print(arr_1[1][1,1])
```

```python
arr_1 = np.array([[[
                   [12,13,14],
                   [34,35,36],
                   [75,76,77]],
                  
                  [[89,88,48],
                   [6,67,89],
                   [87,85,86]]],
                 [[
                   [22,23,24],
                   [24,25,26],
                   [25,26,27]],
                  
                  [[42,43,44],
                   [44,45,46],
                   [55,46,47]]]])

print(arr_1)

# dimensions
print(np.ndim(arr_1))
# check size 
print(arr_1.size)
# check column and row
print(arr_1.shape) # (section_of array,no_of_array,row,col)

print(arr_1[0][1][2,2])
```
