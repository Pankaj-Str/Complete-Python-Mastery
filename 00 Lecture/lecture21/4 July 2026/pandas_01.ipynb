```python
import pandas as pd

print(pd.__version__)
```

```python
# Series

data = ["java","C++","C#","JSP","go","Pandas","Python"]

for i in data:
    print(i)   
```

```python
# print Series 

for i in range(len(data)):
    print(f"{i} . {data[i]}") 
```

```python
# using pandas 
s = pd.Series(data)
print(s)
```

```python
# update series 
indexx = [101,102,103,104,105,106,107]
s = pd.Series(data,index=indexx)
print(s)
```

```python
# From a dictionary

data = {
    "name":"joy",
    "age":56,
    "height":5.6,
    "city":"mumbai"
}

s = pd.Series(data)
print(s)
```

```python
# DataFrame

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'London', 'Paris', 'Tokyo']
}

df = pd.DataFrame(data)
df
```

```python
# export 
df.to_csv("data.csv")
```
