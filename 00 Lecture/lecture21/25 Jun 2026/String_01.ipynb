```python
# string example 

data = "codeswithpankaj"
print(data)
```

```python
# access characters using index
print(data[5])

# slicing
print(data[5:9])
print(data[:5])
print(data[5:])
print(data[:])
```

```python
# 1. upper() converts the string to uppercase
print(data.upper())
```

```python
# 2. lower() converts the string to lowercase
print(data.lower())
```

```python
# 3. partition() returns a tuple
print(data.partition("with"))
```

```python
# 4. replace() replaces substring inside
print(data.replace("with", "by"))
```

```python
# 5. find() returns the index of first occurrence of substring
print(data.find("w"))
```

```python
# 6. rstrip() removes trailing characters
text = "Hello, World!   "
print(text,"Python")
print(text.rstrip()+"Python")
```

```python
# 7. split() splits string from left
data = "www.codeswithpankaj.com"
print(data.split("."))
```

```python
# 8. startswith() checks if string starts with the specified string
print(data.startswith("www"))
```

```python
# 9. isnumeric() checks numeric characters
number = "12345"
print(number.isnumeric())
```

```python
# 10. index() returns index of substring	
print(data.index("pankaj"))		
```

```python
# 11. format() formats the string

name = "nishant"
age = 25
height = 5.9

print("my name is ",name)
print("my age is ",age)
print("my height is ",height)

# using format() method
print("my name is {1} and my age is {0} and my height is {2}".format(name,age,height))

print(f"my name is {name} and my age is {age} and my height is {height}")
```
