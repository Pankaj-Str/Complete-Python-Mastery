```python
import matplotlib.pyplot as plt

# Line charts

sales = [23,34,56,20,56,78,90,100,34,56]
year = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]

plt.plot(year,sales,color='red', linestyle='--', marker='o')
plt.title("Sales year")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()

# Bar charts
# Pie charts
# Scatter plots
# Histograms
```

```python
# bar chart 
plt.bar(year,sales)
plt.title("Sales year")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
# pie chart 

labels = ["Python", "Java", "C++"]
sizes = [40, 35, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Language Usage")
plt.show()
```

```python
plt.scatter(year,sales)
plt.title("Sales year")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
plt.hist(sales)
plt.title("Sales year")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
x = [1, 2, 3, 4]

y1 = [10, 20, 30, 40]
y2 = [40, 30, 20, 10]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.legend()
plt.show()
```

```python
students = ["A", "B", "C", "D"]
marks = [75, 85, 60, 90]

plt.bar(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
