```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

print(sns.__version__)
```

```python
# Optional: Set a beautiful default style (highly recommended)
# https://seaborn.pydata.org/tutorial/color_palettes.html
sns.set_theme(style="whitegrid", palette="deep")
plt.rcParams['figure.figsize'] = (8, 5)  # Default figure size
```

```python
# load dataset 
tips = sns.load_dataset("tips")
tips.head()
```

```python
print(tips.shape)
```

```python
print(tips.info())
```

```python
# Relational Plots (Relationship between two numeric variables)

sns.scatterplot(data=tips,
                x="total_bill",
                y="tip",
                hue="sex",
                style="smoker",
                size="size",
                palette="deep")
plt.title("Relational Plots (Relationship between two numeric variables)")
plt.show()
```

```python
data = sns.load_dataset("flights")
data.head()
```

```python
sns.lineplot(data=data,y="passengers",x="year",hue="month")
plt.show()
```

```python
# bar plot 
import numpy as np
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", estimator=np.mean)
plt.title("Average Total Bill by Day and Sex")
plt.show()
```
