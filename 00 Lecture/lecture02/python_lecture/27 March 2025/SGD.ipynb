```python
import numpy as np 
from sklearn.linear_model import SGDRegressor
import matplotlib.pyplot as plt
```

```python
# Hours	Scores
# 2	30
# 3	32
# 5	25
# 7	40
# 2	35
# 8	80
# 8	75
# 8	80
```

```python
# Our data
hours = np.array([2,3,5,7,2,8,8,8]).reshape(-1, 1)  # Study hours
scores = np.array([30,32,25,40,35,80,75,80])  # Test scores
```

```python
# make model 
model = SGDRegressor(learning_rate="constant",eta0=0.01, max_iter=100 , random_state=42)
model.fit(hours,scores)

# line details
beta_1 = model.coef_[0]
beta_0 = model.intercept_[0]
print(f"beta 1 {beta_1:2f}")
print(f"beta 0 {beta_0:2f}")
```

```python
#guess all score
gussed_score = model.predict(hours)

plt.scatter(hours,scores,color='blue',label='Real Score')
plt.plot(hours,gussed_score,color='red',label="Our Line")
plt.xlabel("Study Hour")
plt.ylabel("Test Scores")
plt.title("SGD SCikit learn")
plt.legend()
plt.show()
```
