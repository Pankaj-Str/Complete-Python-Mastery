```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("titanic")
df.head()
```

```python
print(df.shape)
```

```python
print(df.describe())
```

```python
sns.pairplot(df,hue="sex" , diag_kind="kde")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidths=0.5)
plt.show()
```

```python
sns.histplot(df["age"],bins=30,kde=True)
plt.show()
```

```python
sns.histplot(df["fare"],bins=100,kde=True)
plt.show()
```

```python
sns.boxplot(x='pclass',y='age',data=df)
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Age Class")
plt.show()
```

```python
sns.violinplot(x='survived',y='age',data=df,hue='sex',split=True)
plt.xlabel("survived")
plt.ylabel("Age")
plt.title("survived Class")
plt.show()
```

```python
sns.countplot(x='sex',hue='survived',data=df)
plt.xlabel("Gender")
plt.ylabel("count")
plt.title("Gender and survived")
plt.legend(title="survived",loc="upper right", labels=['Yes','No'])
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=df,hue='survived',style='sex')
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Age vs fare')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=df,hue='pclass')
plt.show()
```
