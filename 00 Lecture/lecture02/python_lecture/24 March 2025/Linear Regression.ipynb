```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# data set
X = np.array([8,10,12]).reshape(-1,1)
Y = np.array([10,13,16])

# model 
model = LinearRegression()
model.fit(X,Y)

# get slope(m)

m = model.coef_[0] # slope
b = model.intercept_

print(f"slope (m) {m}")
print(f"intercept (b) {b}")


# predict price 20 inch  pizza price

X_new = np.array([20]).reshape(-1,1)
Y_pred = model.predict(X_new)
print(f"Predicted price 20 inch pizza = ${Y_pred[0]:2f}")
```

```python
#plot 
plt.scatter(X,Y, color="blue",label="Data Point")
plt.plot(X,model.predict(X),color="red" ,label="Regression line")
plt.xlabel("Diameter (inches)")
plt.ylabel("$ price")
plt.title("Pizza price Prediction")
plt.legend()
plt.show()
```
