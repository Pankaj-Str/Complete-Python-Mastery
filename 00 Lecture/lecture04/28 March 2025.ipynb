```python
# example 01

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a data set
data = {
    'x': np.linspace(0, 10, 100),
    'y': np.sin(np.linspace(0, 10, 100))
}
df = pd.DataFrame(data)
print(df.head())
```

```python
# create a plot
plt.plot(df['x'], df['y'])
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
# example 02
# create a scatter plot

# create a data set

# student marks
# study hours

data = {
    'study_hours': np.random.randint(1, 10, 100),
    'marks': np.random.randint(50, 100, 100)
}
df = pd.DataFrame(data)
print(df.head()) 
```

```python
# create a scatter plot
plt.scatter(df['study_hours'], df['marks'])
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid()
plt.show()
```

```python
# create a bar plot
# create a data set
# sales data
data = {
    'product': ['A', 'B', 'C', 'D'],
    'sales': [100, 200, 150, 300]
}
df = pd.DataFrame(data)
print(df.head())

# create a bar plot
plt.bar(df['product'], df['sales'])
plt.title('Sales Data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid()
plt.show()


```

```python
# change color each bar

plt.bar(df['product'], df['sales'], color=['red', 'blue', 'green', 'orange'])
plt.title('Sales Data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid()
plt.show()

```

```python
# Histogram
# create a data set

# age data
data = {
    'age': np.random.randint(18, 60, 100)
}
df = pd.DataFrame(data)
print(df.head())

# ceate a histogram
plt.hist(df['age'], bins=10, color='blue', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Pie Chart

# create a data set
# sales data
data = {
    'product': ['A', 'B', 'C', 'D'],
    'sales': [100, 200, 150, 300]
}
df = pd.DataFrame(data)
print(df.head())

# create pie chart

plt.pie(df['sales'], labels=df['product'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Data')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Box Plot
# create a data set

# age data
data = {
    'age': np.random.randint(18, 60, 100)
}
df = pd.DataFrame(data)
print(df.head())

# create a box plot

plt.boxplot(df['age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Heatmap

# create a data set
# correlation matrix
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10)
}
df = pd.DataFrame(data)
print(df.head())


# create a heatmap

plt.imshow(df.corr(), cmap='hot', interpolation='nearest')
plt.title('Correlation Matrix')
plt.colorbar()
plt.xticks(range(len(df.columns)), df.columns, rotation=45)
plt.yticks(range(len(df.columns)), df.columns)
plt.show()
```
