```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Pankaj-Str/Complete-Python-Mastery/refs/heads/main/39%20Day%20MATLAB%20Basic/01%20From%20MATLAB%20to%20Python/Basic_Example/HistoricalPrices.csv')

print(data.head())
```

```python
data = data.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
print(data.info())
```

```python
# convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])


```

```python
print(data.info())
```

```python
data = data.sort_values(by = 'Date')
```

```python
print(data.head())
print(data.tail())
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
plt.plot(data['Date'], data['Close'])
plt.show()
```

```python
plt.plot(data['Date'], data['Open'], label='Open')
plt.plot(data['Date'], data['Close'], label='Close')
plt.plot(data['Date'], data['High'], label='High')
plt.plot(data['Date'], data['Low'], label='Low')
plt.title('Historical Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Import the calendar package 
from calendar import month_name
# Order by months by chronological order
data['Month'] = pd.Categorical(data['Date'].dt.month_name(), month_name[:])

# Group metrics by monthly averages
monthly_mean = data \
    .groupby('Month') \
    .mean() \
    .reset_index()

monthly_mean.head(6)
```

```python
# show the monthly mean of 'Close' prices`
plt.plot(monthly_mean['Month'], monthly_mean['Close'])
plt.title('Monthly Average Close Prices')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# create a bar plot for monthly average close prices
plt.bar(monthly_mean['Month'], monthly_mean['Close'])
plt.title('Monthly Average Close Prices (Bar Plot)')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# create horizontal bar plot for monthly average close prices
plt.barh(monthly_mean['Month'], monthly_mean['Close'])
plt.title('Monthly Average Close Prices (Horizontal Bar Plot)')
plt.xlabel('Average Close Price')
plt.ylabel('Month')
plt.show()
```

```python
# scatter plot for monthly average close prices
plt.scatter(monthly_mean['Month'], monthly_mean['Close'])
plt.title('Monthly Average Close Prices (Scatter Plot)')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Data Manipulation
data['Price'] = data['Close'] * 1.1  # Example manipulation
print(data.head())
# Save the manipulated data to a new CSV file
data.to_csv('manipulated_data.csv', index=False)
```
