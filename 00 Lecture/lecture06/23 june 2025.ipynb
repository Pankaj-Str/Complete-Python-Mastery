```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# set the style of seaborn
sns.set_theme(style='darkgrid')
# option to set the style to whitegrid, dark, white, or ticks
```

```python
# load sample dataset
tips = sns.load_dataset('tips')
# display the first few rows of the dataset
print(tips.head())
```

```python
# create a basic scatter plot

# example = plot total bill vs tip to see if higher bills lead to higher tips
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')

# add title and labels
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# show the plot
plt.show()
```

```python
# line plot example
# example = create a synthetic dataset to show a trend over time

# create a sample data 
data = pd.DataFrame({"month": range(1, 13),
                     "sales": [200, 220, 250, 275, 300, 320, 350, 400, 450, 500, 550, 600]})
```

```python
# cerate a line plot
sns.lineplot(data=data, x='month', y='sales', marker='o')

# customize the plot
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
# show the plot
plt.show()
```

```python
# create bar plot example
# compare average tips by day of the week
sns.barplot(x='day', y='tip', data=tips)

# customize the plot
plt.title('Average Tip by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip ($)')
# show the plot
plt.show()
```

```python
# box plot example
# exmaple = show the distribution for tips by day
sns.boxplot(x='day', y='tip', data=tips)
# customize the plot
plt.title('Tip Distribution by Day of the Week')

plt.xlabel('Day of the Week')
plt.ylabel('Tip ($)')
# show the plot
plt.show()
```

```python
# histogram example
# example = show the distribution of total bills
sns.histplot(data=tips, x='total_bill', bins=20)
# customize the plot
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
# show the plot
plt.show()
```

```python
# adding hue for grouping
# example = scatter plot total_bill vs tips , colored by sex
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
# customize the plot
plt.title('Total Bill vs Tip by Sex')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
# show the plot
plt.show()
```

```python
# faceting with sns.catplot or sns.relplot
# example = create a scatter plot of total_bill vs tip, faceted by day
sns.catplot(x ='day', y='tip', data=tips, hue='time', kind='bar')
# customize the plot
plt.title('Average Tip by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip ($)')
# show the plot
plt.show()
```

```python
# pairplot example (sns.pairplot)
# example = visualize relationships between all numerical variables in the tips dataset
sns.pairplot(tips, hue='sex')
# customize the plot
plt.suptitle('Pairplot of Tips Dataset', y=1.02)
# show the plot
plt.show()
```

```python
# change the color palette
sns.set_palette('pastel')
# example = create a bar plot with the new color palette
sns.barplot(x='day', y='tip', data=tips)
# customize the plot
plt.title('Average Tip by Day of the Week (Pastel Palette)')
plt.xlabel('Day of the Week')

plt.ylabel('Average Tip ($)')
# show the plot
plt.show()
```

```python
# change the size of the plot
plt.figure(figsize=(10, 6))
# example = create a scatter plot with a larger size
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
# customize the plot
plt.title('Total Bill vs Tip by Sex')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
# show the plot
plt.show()
```

```python
# save the plot to a file
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
# customize the plot
plt.title('Total Bill vs Tip by Sex')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
# save the plot to a file
plt.savefig('total_bill_vs_tip_by_sex.png', dpi=300, bbox_inches='tight')
# show the plot
plt.show()
```
