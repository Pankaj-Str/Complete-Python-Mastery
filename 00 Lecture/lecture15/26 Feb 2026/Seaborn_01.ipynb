```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# change theme
sns.set_style('darkgrid')
```

```python
# load the tips dataset

data = sns.load_dataset('tips')
print(data.head()) # top 5 rows of the dataset

# other data 
# data1 = sns.load_dataset('flights')
# print(data1.head()) # top 5 rows of the dataset
# https://seaborn.pydata.org/tutorial/data_structure.html

# offline dataset
new_data = pd.read_csv('mpg.csv')
print(new_data.head()) # top 5 rows of the dataset
```

```python
# scatter plot of total bill vs tip
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
# add color 
sns.scatterplot(x='total_bill', y='tip', data=data , hue='sex')
plt.show()
```

```python
# line plot of total bill vs tip
sns.lineplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
# bar plot of total bill vs day
sns.barplot(x='day', y='total_bill', data=data)
plt.show()
```

```python
# histogram of total bill
sns.histplot(data['total_bill'], bins=70)
plt.show()
```

```python
sns.histplot(data['tip'], bins=70)
plt.show()
```

```python
# box plot of total bill vs day -> outliers distribution
sns.boxplot(x='day', y='total_bill', data=data)
plt.show()
```

```python
# heatmap of total bill vs tip
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True , cmap='coolwarm')
plt.show()
```
