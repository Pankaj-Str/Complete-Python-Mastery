```python
# download dataset 
# https://www.wsj.com/market-data/quotes/index/DJIA/historical-prices
```

```python
import pandas as pd

# read the dataset
df = pd.read_csv('data.csv')
# display the first 5 rows of the dataset
df.head()
```

```python
# find info 
df.info()
```

```python
df.columns
```

```python
# rename columns
df = df.rename(columns={' Open':'Open',' High':'High',' Low':'Low',' Close':'Close'})
```

```python
# check columns again
df.columns
```

```python
# convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# check info again
df.info()
```

```python
# import matplotlib
import matplotlib.pyplot as plt

# create a line plot of the closing price over time

plt.plot(df['Date'], df['Close'])
plt.show()
```

```python
plt.plot(df['Date'], df['Open'])
plt.plot(df['Date'], df['High'])
plt.plot(df['Date'], df['Low'])
plt.plot(df['Date'], df['Close'])
plt.show()
```

```python
# Import the calendar package 
from calendar import month_name
# Order by months by chronological order
df['Month'] = pd.Categorical(df['Date'].dt.month_name(), month_name[:])

# Group metrics by monthly averages
monthly = df \
    .groupby('Month') \
    .mean() \
    .reset_index()

monthly.head(12)
```

```python
# create a bar plot of the average closing price by month
plt.bar(monthly['Month'], monthly['Close'])
plt.xticks(rotation=90)
plt.show()

# horizontal bar plot
plt.barh(monthly['Month'], monthly['Close'])
plt.show()
```

```python
# scatter plot of open vs close
plt.scatter(df['Open'], df['Close'])
plt.show()
```

```python
# bar change color
plt.bar(monthly['Month'], monthly['Close'], color='red')
plt.xticks(rotation=90)
plt.show()

# different color for each bar
plt.bar(monthly['Month'], monthly['Close'], color=['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta', 'yellow', 'black'])
plt.xticks(rotation=90)
plt.show()
```

```python
# change color of scatter plot
plt.scatter(df['Open'], df['Close'], c=df['Close'], cmap = plt.cm.plasma)
plt.show()
```
