```python
from scipy import constants
print(constants.pi)
print(constants.g) # gravitational acceleration
print(constants.c) # speed of light
print(constants.speed_of_light) # speed of light
```

```python
from scipy import  integrate

def func(x):
    return x**2


result = integrate.quad(func, 0, 1)
print(result)
```

```python
new_data = constants.pi

result = func(78)

print(result/new_data)
```

```python
from scipy import stats

data = [21,23,24,25,26,27,28,29,30]

# print(stats.describe(data))

print(stats.describe(data, ddof=1))

# find mean 
print(stats.tmean(data))

print(stats.mode(data))
```
