```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
points = np.array(
    [
        [1,1],
        [1.5,1.5],
        [5,5],
        [6,6],
        [8,1],
        [8,2],
        [10,10],
        [25,25],
        [2.3,2.3],
        [2.4,2.4],
        [2.5,2.5],
        [2.6,2.6],
        [2.7,2.7],
        [2.8,2.8],
        [60,60]
    ]
)
```

```python
# apply DBSCAN
db = DBSCAN(eps=1.5, min_samples=2).fit(points)
# eps = maximum distance between two samples for them to be considered as in the same neighborhood
# min_samples = number of samples in a neighborhood for a point to be considered as a core point
```

```python
# get labels of clusters 
# -1 means noise points
labels = db.labels_
print(labels)
```

```python
# visualize clusters
unique_labels = set(labels)
colors = ['r','b','g','y'] +['k'] # color for noise points and clusters
for k , col in zip(unique_labels,colors[:len(unique_labels)]):
    if k == -1:
        col = 'k' # black for noise 
    class_member_mask = (labels == k)
    xy = points[class_member_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
    'o', 
    markerfacecolor=col ,
    markeredgecolor ='k',
    markersize = 10,
    label=f"cluster {k}" if k != -1 else "Noise"
    )
plt.title("DBSCAN clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
    
```
