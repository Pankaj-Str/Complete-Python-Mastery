```python
from sklearn.linear_model import Ridge

X = [[1], [2], [3], [4], [5]] # study hours
y = [35, 50, 65, 80, 95] # corresponding scores

model = Ridge(alpha=1.0) # create Ridge regression model with regularization strength alpha
model.fit(X, y) # fit the model to the data

predicted_score = model.predict([[6]]) # predict score for 6 study hours
print(predicted_score) # print the predicted score

print("Coefficients:", model.coef_) # print the coefficients of the model
print("Intercept:", model.intercept_) # print the intercept of the model
```

```python
# lasso regression example
from sklearn.linear_model import Lasso

model_lasso = Lasso(alpha=0.1) # create Lasso regression model with regularization strength alpha
model_lasso.fit(X, y) # fit the Lasso model to the data

predicted_score_lasso = model_lasso.predict([[6]]) # predict score for 6 study hours using Lasso regression
print(predicted_score_lasso) # print the predicted score from Lasso regression

print("Lasso Coefficients:", model_lasso.coef_) # print the coefficients of the Lasso model
print("Lasso Intercept:", model_lasso.intercept_) # print the intercept of the
```
