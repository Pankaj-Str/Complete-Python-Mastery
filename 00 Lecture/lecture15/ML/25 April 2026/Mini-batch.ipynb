```python
# step import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
```

```python
# create a sample dataset
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000, # number of samples
                  centers=4, # number of centers(clusters/groups)
                  cluster_std=0.60,  # standard deviation of clusters
                  random_state=42) # random state for reproducibility
```

```python
# apply mini-batch k-means
kmeans = MiniBatchKMeans(n_clusters=4, # number of clusters(groups)
                         batch_size=20, # number of samples to be used in each iteration
                         random_state=42) # random state for reproducibility
kmeans.fit(X) # fit the model to the data
```

```python
# get cluster labels
labels = kmeans.labels_
print(labels)
```

```python
# visualize the clusters
plt.scatter(X[:, 0], # first feature
            X[:, 1], # second feature
            c=labels,  # color by cluster labels
            cmap='viridis',# color map for better visualization
            marker='o',
            s=10) 

plt.scatter(kmeans.cluster_centers_[:, 0], # cluster centers first feature
            kmeans.cluster_centers_[:, 1], # cluster centers second feature
            c='red', # color for cluster centers
            marker='X', # marker for cluster centers
            s=50, # size of cluster centers
            label='Centroids') # label for legend

plt.title('Mini-Batch K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
