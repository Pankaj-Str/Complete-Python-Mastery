```python
# load dataset 
import seaborn as sns

# load the penguins dataset 
data = sns.load_dataset('penguins')

# show data 
data.head()
```

```python
# shape
data.shape
```

```python
# clean data 
# remove rows with missing values
data = data.dropna()
data.head()
```

```python
# import pandas 
# feature(X) and target(y)
import pandas as pd

X = data[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
y = data['species']

print('Feature = shape (x) ',X.shape)
print('target = shape (y) ',y.shape)
```

```python
# split data into Traning and Testing 
from sklearn.model_selection import train_test_split

# 20% testing
# 80% traning

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.5,random_state=42)

# check data 
print('Traning Data Features size : ',X_train.shape)
print('Testing Data Features size : ',X_test.shape)

print('Traning Data target size : ',y_train.shape)
print('Testing Data target size : ',y_test.shape)
```

```python
# create a train model - random forest model 
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=10,random_state=42)

# train the model 
rf_model.fit(X_train , y_train)
```

```python
# make predictions 
y_pred = rf_model.predict(X_test)

# print actual answer 
print('answer actual = ',y_pred[:5])
# predictions answer 
print('answer actual = ',y_test[:5].values)
```

```python
# check accuracy_score 
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)

print(f'accuracy score {accuracy*100:.2f}%')
```

```python
# check test _ new penguin

new_penguin = [[40.1,18.0,190,4000]]
# make predictions 
prediction = rf_model.predict(new_penguin)

print(f'Predicted = {prediction[0]}')
```
