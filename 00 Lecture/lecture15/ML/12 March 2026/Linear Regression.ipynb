```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# dataset 
X = np.array([8,10,12]).reshape(-1,1) # size
y = np.array([10,13,16]) # price 
```

```python
# model - LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
# find (m) slope and intercept (b)
m = model.coef_  # (m) slope
b = model.intercept_ # intercept (b)
print(f"slope {m}")
print(f"intercept {b}")
```

```python
# predict 
# 20 pizza price ?
X_new = np.array([20]).reshape(-1,1)
y_pred = model.predict(X_new)
print(f"predicted price 20 inch pizza = {y_pred}")
```
