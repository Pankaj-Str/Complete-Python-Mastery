```python
# import lib
from sklearn.tree import DecisionTreeClassifier , plot_tree , export_text
import matplotlib.pyplot as plt
```

```python
X = [
    [25,50000],
    [30,60000],
    [45,80000],
    [55,95000]
]

y = [0,0,1,1]
```

```python
# create a model 
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
tree_rules = export_text(model,feature_names=["Age","Income"])
print(tree_rules)
```

```python
#plot 
plt.Figure(figsize=(12,8))
plot_tree(model,feature_names=["Age","Income"],class_names=["YES","NO"],filled=True,rounded=True,fontsize=10)
plt.show()
```
