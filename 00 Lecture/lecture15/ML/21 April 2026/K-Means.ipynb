```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# customer data
customers = np.array([
    [20,500], # Rohan
    [25,600], # Priya
    [30,700], # Ankit
    [35,800], # Neha
    [40,900], # Rahul
])
```

```python
# run K-Means with 2 groups
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(customers)
```

```python
# get group assignments and centers for each group
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Draw the clusters and centers
plt.scatter(customers[:, 0], customers[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200)
plt.xlabel('Age')
plt.ylabel('Annual Spending')
plt.title('Customer Segmentation')
plt.show()
```

```python
# how pick the right number of group - Elbow method
wcss = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(customers)
    wcss.append(kmeans.inertia_)
    
# drow the elbow plot
plt.plot(range(1, 6), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.show()    
```
