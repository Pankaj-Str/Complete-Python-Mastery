```python
import math


def sigmoid(z):
    return 1 / (1 + math.exp(-z))
```

```python
hours  = 4

z = -1.5 + 0.6 * hours

print(z)
```

```python
chance = sigmoid(z)

print(chance)
print(f"Change = {chance*100:.2f}%")
```

```python
predictipon = "pass" if chance > 0.5 else "fail" 
print(predictipon)
```
