```python
import numpy as np
print(np.__version__)
```

```python
# create a single dimensional array
data = [21,22,23,24,25]
print(data)
# check type 
print(type(data))

# access element
print(data[0])
print(data[1])
print(data[2])
```

```python
data = np.array([21,22,23,24,25])
print(data)
# check type 
print(type(data))

# check dimension
print(data.ndim)
# shape of array
print(data.shape)
# size of array
print(data.size)

# access element
print(data[0])
print(data[1])
print(data[2])
```

```python
# 2 dimensional array
data = np.array([[21,22,23],[24,25,26],[27,28,29]])
print(data)
# check dimension
print(data.ndim)
# shape of array
print(data.shape)
# size of array
print(data.size)

# access element
print(data[0][0])
print(data[2][1])
print(data[0][2])
```

```python
# 3 dimensional array
data = np.array([[
    [21,22,23],
    [24,25,26],
    [27,28,29]],
                 [[2,3,53],
                  [54,15,76],
                  [77,98,89]]])
print(data)
# check dimension
print(data.ndim)
# shape of array
print(data.shape)
# size of array
print(data.size)

# access element
print(data[0][0][0])

print(data[1][0][0])
```

```python
# 4 dimensional array
data = np.array([[[
    [21,22,23],
    [24,25,26],
    [27,28,29]],
                 [[21,22,23],
                  [24,25,26],
                  [27,28,29]]],
                [[
    [21,22,23],
    [24,25,26],
    [27,28,29]],
                 [[21,22,23],
                  [24,25,26],
                  [27,28,29]]]])
print(data)
# check dimension
print(data.ndim)
# shape of array
print(data.shape)
# size of array
print(data.size)

# access element
print(data[0][0][0][0])

print(data[1][0][0][0])
```
