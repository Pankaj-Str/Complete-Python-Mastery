```python
import math
from collections import Counter
import pandas as pd

doc = [
    "The cat sat on the mat",
    "The dog sat on the mat",
    "The cat and the dog"
]
```

```python
# Lower case split into word
docs = [doc.lower().split() for doc in doc]
print(docs)
```

```python
# find the all unique words

terms = sorted(set(word for doc in docs for word in doc))
print(terms)

N = len(docs)
print("total documents = ",N)
```

```python
# document frequency (DF)
df = {}
for term in terms:
    df[term] = sum(1 for doc in docs if term in doc)

print(df)
```

```python
# Inverse documents frequnecy (IDF)
idf = {term: math.log(N/df) for term, df in df.items()}
print(idf)
```

```python
# build TF - IDF matrix
tfidf_matrix = []

for doc in docs:
    tf = Counter(doc)
    tfidf = {term: tf[term] * idf[term] for term in doc}
    tfidf_matrix.append(tfidf)

print(tfidf_matrix)
```

```python
# show in table format
df_tfidf = pd.DataFrame(tfidf_matrix)
df_tfidf.index = ["doc1", "doc2", "doc3"]
df_tfidf.fillna(0, inplace=True)

df_tfidf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(doc)
print(vectorizer.get_feature_names_out())
print(tfidf_matrix.toarray())
```
