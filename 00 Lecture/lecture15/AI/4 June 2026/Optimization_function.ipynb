```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(w):
    return w**2

def gradient(w):
    return 2 * w # derivative of w^2
```

```python
# Gradient Descent (GD) function 
def gradient_descent(start_w,learning_rate,epochs):
    w = start_w
    history = [w] # track update
    for _ in range(epochs):
        grad = gradient(w)
        w = w - learning_rate * grad
        history.append(w)
    return history    
    
```

```python
# start 4:3
start_w = 4.0
learning_rate = 0.1
epochs = 20

update = gradient_descent(start_w,learning_rate,epochs)
print(update[:5])
```

```python
#plot 
plt.plot(range(epochs+1),[cost(w) for w in update],marker='o')
plt.title("Gradient Descent")
plt.xlabel("update")
plt.ylabel("cost (w^2)")
plt.grid(True)
plt.show()
```
