```python
import numpy as np
import matplotlib.pyplot as plt

def mean_squared_error(actual, predicted):
    return np.mean((actual - predicted) ** 2)


# test data (multiple data points)

actual = np.array([3, 5, 2, 7]) # true values
predicted_good = np.array([2.5, 5.0, 2.1, 7.8]) # predicted values -> closer to actual values
predicted_bad = np.array([1.0, 4.0, 3.0, 6.0]) # predicted values -> farther from actual values



print("Mean Squared Error (Good Predictions):", mean_squared_error(actual, predicted_good))
print("Mean Squared Error (Bad Predictions):", mean_squared_error(actual, predicted_bad))


# binary cross entropy loss function
def binary_cross_entropy(actual, predicted):
    # Clip predicted values to avoid log(0)
    predicted = np.clip(predicted, 1e-15, 1 - 1e-15) # avoid log(0) by clipping predicted values to a small range
    return -np.mean(actual * np.log(predicted) + (1 - actual) * np.log(1 - predicted))


# test data for binary classification
actual_binary = np.array([1, 0, 1, 0]) # true binary labels
predicted_binary_good = np.array([0.9, 0.1, 0.8, 0.2]) # predicted probabilities -> closer to actual labels
predicted_binary_bad = np.array([0.2, 0.8, 0.3, 0.7]) # predicted probabilities -> farther from actual labels


print("Binary Cross Entropy (Good Predictions):", binary_cross_entropy(actual_binary, predicted_binary_good))
print("Binary Cross Entropy (Bad Predictions):", binary_cross_entropy(actual_binary, predicted_binary_bad))
```

```python
# visualization of the loss functions

# predicted values for visualization
predicted_values = np.linspace(0, 1, 100) # predicted probabilities from

cost = -np.log(predicted_values) # cost for actual = 1
cost += -np.log(1 - predicted_values) # cost for actual = 0

plt.figure(figsize=(10, 6))
plt.plot(predicted_values, cost, label='Binary Cross Entropy Loss', color='blue')
plt.title('Binary Cross Entropy Loss Function')
plt.xlabel('Predicted Probability')
plt.ylabel('Cost')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0.5, color='red', linestyle='--', label='Threshold (0.5)')
plt.legend()
plt.grid()
plt.show()
```
