```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer , WordNetLemmatizer
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "codes with pankaj is best learning platform! and using python , R programming a lot."
```

```python
# Tokenization: Split text into words or sentences.
words = word_tokenize(text)
print("Tokenized Words:", words)

sentences = nltk.sent_tokenize(text)
# print("Tokenized Sentences:", sentences)

for i in sentences:
    print(i)
```

```python
# Remove Stop Words: "the", "is", "and" (common, low-value words).

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print("Filtered Words:", filtered_words)
```

```python
# Stemming: Reduce words to root (e.g., "playing" → "play").


stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("Stemmed Words:", stemmed_words)
```

```python
# Lemmatization: Better than stemming – considers context (e.g., "better" → "good").
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("Lemmatized Words:", lemmatized_words)
```

```python
# POS tagging
from nltk import pos_tag
pos_tags = pos_tag(filtered_words)
print("POS Tags:", pos_tags)
```
