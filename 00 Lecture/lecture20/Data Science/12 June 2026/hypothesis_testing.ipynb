```python
import numpy as np
from scipy import stats

# sample data 
np.random.seed(42)
sample_marks = np.random.normal(loc=72,scale=8,size=30)

# hypotheses 
# h0 mu = 75
# H1 mu != 75

# one sample T-test 
t_stat , p_value = stats.ttest_1samp(sample_marks,popmean=75)
print("Test Statistic (t) : ",round(t_stat,4))
print("p Values = ",round(p_value,4))
```

```python
alpha = 0.05
if p_value < alpha:
    print("Result : Reject H0  -> Average Marks are significantliy diffrent from 75")
else:
     print("Result : Fail to Reject  -> No significantliy diffrerence from 75")
```
