```python
import pandas as pd

students = pd.DataFrame({
        'StudentID':[101,102,103,104],
        'Name':['joy','rohan','Toy','Roy'],
        'Age':[20,21,19,22]
    })

scores = pd.DataFrame({
    'StudentID':[101,102,103,105],
    'Subject':['Maths','Maths','Science','English'],
    'Score':[85,95,65,78]
})
print(students)
print(scores)
```

```python
join_inner = pd.merge(students,scores,on='StudentID')
join_inner
```

```python
left_join = pd.merge(students,scores,on='StudentID',how='left')
left_join
```

```python
right_join = pd.merge(students,scores,on='StudentID',how='right')
right_join
```

```python
outer_join = pd.merge(students,scores,on='StudentID',how='outer')
outer_join
```
