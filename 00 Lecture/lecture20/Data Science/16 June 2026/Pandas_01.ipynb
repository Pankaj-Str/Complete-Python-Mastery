```python
import pandas as pd

print(pd.__version__)
```

```python
# Series() method

data = ["java","c++","ruby","go","jsp","julia","python"]
# print(data)
for i in data:
    print(i)
```

```python
# create a index 
size = len(data)
for i in range(size):
    print(f"{i} = {data[i]}")
```

```python
# Series() method
s = pd.Series(data)
print(s)
```

```python
indexx = ['a','b','c','d','e','f','g']

s = pd.Series(data,index=indexx)
print(s)
```

```python
data = {
    'a':100,'b':300,'c':400,'d':450
}
#print(data)
s = pd.Series(data)
print(s)
```

```python
import numpy as np
data = np.array([12,13,14,15,16])
s = pd.Series(data)
print(s)
```

```python
data = ["java","c++","ruby","go","jsp","julia","python"]
s = pd.Series(data)
print(s)

# access data from index

print(s[3])

# update index 
indexx = ['a','b','c','d','e','f','g']

s = pd.Series(data,index=indexx)
print(s['c'])
```
