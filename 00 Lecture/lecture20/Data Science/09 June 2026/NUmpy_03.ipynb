```python
# import numpy as np
import numpy as np
from numpy import random

x = random.randint(100)
print(x)

x = random.rand()
print(x)
```

```python
x = random.randint(100,size=(3,5))
print(x)
```

```python
# Float Array 
x = random.rand(5)
print(x)

# create 2d array 
x = random.rand(3,5)
print(x)
```

```python
# single values 
x = random.choice([3,4,5,6,7,8,9,10])
print(x)
```

```python
# array of values 2d array 
x = random.choice([1,2,3,4],size=(3,3))
print(x)
```

```python
color = ['red','green','blue','lightgreen','pink']
selected = np.random.choice(color,size=3)
print(selected)
```

```python
# shuffle an array 
arr = np.array([21,22,23,24,25])
np.random.shuffle(arr)
print(arr)

# permuation
new_arr = np.random.permutation(arr)
print(new_arr)
```

```python
# seed()
np.random.seed(42)
x = np.random.randint(1,100)
print(x)
```
