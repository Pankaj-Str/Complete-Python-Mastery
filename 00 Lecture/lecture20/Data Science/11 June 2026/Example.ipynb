```python
import numpy as np

marks = np.array([60,65,70,75,80])
print("Students Marks")
print(marks)
```

```python
# mean calculate (avg)
mean = np.mean(marks)
print("Average Marks = ",mean)
```

```python
# calculate Standard Deviation
std = np.std(marks)

print("standard Deviation = ",std)

print("standard Deviation = ",round(std,2))
```

```python
# calculate Z-Scores 
z_scores = (marks-mean)/std

print("Z scores = ",z_scores)
```

```python
for mark , z in zip(marks,z_scores):
    print(f"Marks = {mark}, Z score {z:.2f}")
```

```python
# T Score

Sample_mean = np.mean(marks)
population_mean = 65
Sample_sd = np.std(marks,ddof=1)
n = len(marks)

t_score = (Sample_mean - population_mean)/(Sample_sd/np.sqrt(n))
print("\n T score ",round(t_score,2))

#  degrees of freedom
df = n - 1
print("degrees of freedom  = ",df)
```
