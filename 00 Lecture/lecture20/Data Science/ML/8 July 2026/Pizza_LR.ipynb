```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# create a dataset
X = np.array([8,10,12]).reshape(-1,1) # pizza size 
Y = np.array([10,13,16]) # pizza price
```

```python
model = LinearRegression()
model.fit(X , Y)
# slope (m)
m = model.coef_[0]
print(f"Slope = {m}")
# intercept (b)
b = model.intercept_
print(f"Intercept (b) = {b}")
```

```python
# predict price size = 20 inch
X_new = np.array([20]).reshape(-1,1)
Y_pred = model.predict(X_new)
print(f"20 inch pizza price = ${Y_pred[0]:.2f}")
```

```python
# Plot 
plt.scatter(X,Y,color='blue',label='Data Point')
plt.plot(X , model.predict(X), color='red', label="Regression line")
plt.xlabel("inches")
plt.ylabel("Price $")
plt.title("Pizza price Prediction")
plt.legend()
plt.show()
```
