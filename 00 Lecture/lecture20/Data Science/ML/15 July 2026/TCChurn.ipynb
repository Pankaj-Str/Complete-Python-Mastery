```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
# load dataset 
df = pd.read_csv('Telco-Customer-Churn.csv')
df.head()
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
# basic data cleaning 
df = df.drop(['customerID'],axis=1)
# check delete columns 
df.columns
```

```python
# fix TotalCharges column dtypes 
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())

# print(df.isnull().sum())

df['Churn'] = df['Churn'].map({'No':0,'Yes':1})
df.head(2)
# print(df.isnull().sum())
```

```python
# Feature and target
X = df.drop(['Churn'],axis=1)
y = df['Churn']
```

```python
X.head(2)
```

```python
print(df.columns)
```

```python
categorical_cols = ['gender','SeniorCitizen','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']

numerical_cols = ['tenure','MonthlyCharges', 'TotalCharges']

# preprocessing 

preprocessor = ColumnTransformer(
    transformers=[
        ('num',StandardScaler(),numerical_cols),
        ('cat',OneHotEncoder(drop='first',sparse_output=False),categorical_cols)
])
```

```python
# train test split
# test 30%
# train 70%

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=42)

print('Train data X ',X_train.shape)
print('Train data Y ',y_train.shape)

print('Test data X ',X_test.shape)
print('Test data Y ',y_test.shape)


# Create a model pipeline 
model = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('classifier',LogisticRegression(max_iter=1000,random_state=42))
])

# train the model 
model.fit(X_train,y_train)
```

```python
# make predictions 
y_pred = model.predict(X_test)
print("Accuracy : \n",accuracy_score(y_test,y_pred))
print("classification report : \n",classification_report(y_test,y_pred))
print("confusion matrix \n",confusion_matrix(y_test,y_pred))
```
