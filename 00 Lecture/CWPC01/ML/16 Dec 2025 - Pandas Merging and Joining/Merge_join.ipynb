```python
# inner join 
inner_join = pd.merge(employee,departments,on='Dept_id',how='inner')
inner_join
```

```python
# left join 
left_join = pd.merge(employee,departments,on='Dept_id',how='left')
left_join
```

```python
# right join 
right_join = pd.merge(employee,departments,on='Dept_id',how='right')
right_join
```

```python
# outer join 
outer_join = pd.merge(employee,departments,on='Dept_id',how='outer')
outer_join
```
