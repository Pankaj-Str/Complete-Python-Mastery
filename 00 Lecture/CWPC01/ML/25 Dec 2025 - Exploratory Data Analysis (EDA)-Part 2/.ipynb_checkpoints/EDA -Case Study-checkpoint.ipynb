```python
import pandas as pd 
url = "https://archive.ics.uci.edu/static/public/10/data.csv"
url2 = "http://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url2)
print(df)
```

```python
# column name 
columns = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
data = pd.read_csv(url2,names=columns,na_values="?",sep=",",skipinitialspace=True)
data.head()
```

```python
data_unique = ['make', 'fuel-type', 'aspiration','num-of-doors', 'body-style', 'drive-wheels', 'engine-location','num-of-cylinders','fuel-system','engine-type']
for col in data_unique:
    print(f'{col} : {data[col].unique()}')
```

```python
# handel missing values 
data.isna().sum()
```

```python
normalized_losses = data['normalized-losses'].mean()
data['normalized-losses'].fillna(normalized_losses,inplace=True)
data['normalized-losses'].head()
```

```python
# handel missing values 
data.isna().sum()
```

```python
price = data['price'].mean()
data['price'].fillna(price,inplace=True)
data['price'].head()

bore = data['bore'].mean()
data['bore'].fillna(bore,inplace=True)
data['bore'].head()
```

```python
# handel missing values 
data.isna().sum()
```

```python
horsepower = data['horsepower'].mean()
data['horsepower'].fillna(horsepower,inplace=True)

stroke = data['stroke'].mean()
data['stroke'].fillna(stroke,inplace=True)

peak_rpm = data['peak-rpm'].mean()
data['peak-rpm'].fillna(peak_rpm,inplace=True)
```

```python
data.isna().sum()
```

```python
data['num-of-doors'] = data['num-of-doors'].fillna("Unknown")
```

```python
data.isna().sum()
```

```python
data.to_csv("auto_mobile.csv")
```

```python
data.describe()
```

```python
# IQR
q1 = data['horsepower'].quantile(0.25)
q3 = data['horsepower'].quantile(0.75)

iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliner = data[(data['horsepower'] < lower) | (data['horsepower'] > upper)]
print(outliner[['make','horsepower']].head())
print(upper)
print(lower)
print(iqr)
```

```python
# % pip install scipy
from scipy import stats
import numpy as np

data['zscore'] = stats.zscore(data['horsepower'])
z_outliners = data[np.abs(data['zscore']) > 3]
print(z_outliners[['make','horsepower']].head())
```
