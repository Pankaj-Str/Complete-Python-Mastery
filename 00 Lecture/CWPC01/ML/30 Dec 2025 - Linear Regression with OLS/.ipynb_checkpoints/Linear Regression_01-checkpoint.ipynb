```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# dataset 
X = np.array([8,10,12]).reshape(-1,1) # pizza diameters
Y = np.array([10,13,16]) # price 

# create a model 
model = LinearRegression()
model.fit(X,Y) # put values into model
```

```python
# get slope (m) intercept (b)
m = model.coef_[0] # Slope
b = model.intercept_ # intercept

print(f"Slope m = {m}")
print(f"intercept b = {b} ")
```

```python
# predict price = 20 inch
X_new = np.array([20]).reshape(-1,1)
Y_pred = model.predict(X_new)
print(f"predicted price  for 20 inch pizza = {Y_pred[0]:.2f}")
```

```python
plt.scatter(X,Y,color="blue",label="Data point")
plt.plot(X,model.predict(X),color="red",label="Regression line")
plt.scatter(X_new,Y_pred,color="green",label="Regression line new")
plt.xlabel('Diameter (inche)')
plt.ylabel('Price $')
plt.title('Pizza Price Predication')
plt.legend()
plt.show()
```
