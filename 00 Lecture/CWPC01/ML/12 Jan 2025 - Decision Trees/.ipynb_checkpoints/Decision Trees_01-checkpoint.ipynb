```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier , plot_tree
```

```python
# create dataset 
X_custom = [[5,1],[5,0],[2,1],[1,0]] # Features : [Hours Studied , sleep]
y_custom = [1,0,0,0] #labels : pass 1 fail 0
```

```python
# initialize and train data
clf_custom = DecisionTreeClassifier(criterion="gini",random_state=42)
clf_custom.fit(X_custom,y_custom)
```

```python
plot_tree(
    clf_custom,
    feature_names=['Hours','sleep'],
    class_names = ['fail','pass'],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
# load dataset
iris = load_iris()
X , y = iris.data , iris.target
```

```python
# slipt data 
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42)
# train the tree
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
plt.figure(figsize=(10,8))
tree.plot_tree(clf,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,fontsize=7)
plt.show()
```
