```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
# house vs price 

X = np.array([1.0,1.5,2.0,3.0,4.0,5.0]).reshape(-1,1)
y = np.array([15,18,22,30,45,70])
```

```python
svr_linear = SVR(kernel='linear', C = 100 ,epsilon=3)
svr_poly = SVR(kernel='poly', degree=3 , C=100 , epsilon=3,coef0=1)
svr_rbf = SVR(kernel='rbf' ,  C=100 , gamma=0.5 , epsilon=3)
```

```python
svr_linear.fit(X,y)
svr_poly.fit(X,y)
svr_rbf.fit(X,y)
```

```python
# predict on smooth points curve
X_plot = np.linspace(0.5,5.5,200).reshape(-1,1)
y_linear = svr_linear.predict(X_plot)
y_poly = svr_poly.predict(X_plot)
y_rbf = svr_rbf.predict(X_plot)
```

```python
plt.scatter(X,y, color='red',label='Actual data ',zorder=10)
plt.plot(X_plot,y_linear , label='linear kernel ', linewidth=2)
plt.plot(X_plot,y_poly , label='polynomial kernel ', linewidth=2)
plt.plot(X_plot,y_rbf , label='RBF kernel ', linewidth=2)

plt.title('SVR')
plt.xlabel('size')
plt.ylabel('Price')
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```
