```python
# https://editor.analyticsvidhya.com/uploads/51518iris%20img1.png
# https://scikit-learn.org/stable/datasets/toy_dataset.html
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris , load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , classification_report
```

```python
# load data set
iris = load_iris()
print(iris)
```

```python
# select columns 

print(iris.data)
print(iris.target_names)
print(iris.feature_names)
print(iris.target)
```

```python
# convert into DataFrame
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# iris.target
df['target'] = iris.target

df.head()
```

```python
# feature name (input)
X = df.drop('target',axis=1)
X.head()
```

```python
# target (output)
y = df['target']
```

```python
# split data into train data
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.1,random_state=42)

# 80%  data -> training
# 20% data -> testing
```

```python
# create a random forest model
rf_model = RandomForestClassifier(n_estimators=100,random_state=42)
# n_estimators=100 -> number of tree 
```

```python
# train the model
rf_model.fit(X_train,y_train)
```

```python
# make predict
y_pred = rf_model.predict(X_test)
```

```python
# check model accuracy
accuracy = accuracy_score(y_test,y_pred)
print('accuracy score : ',accuracy)
```

```python
# details classifications report 
print(classification_report(y_test,y_pred))
```

```python
# predict for new data 
# new flower data 
new_flower = [[5.1,3.5,1.4,0.2]]
prediction = rf_model.predict(new_flower)
print('prediction flower type ',iris.target_names[prediction][0])
```

```python
# visualization 1 Feature importance 
import matplotlib.pyplot as plt
importance = rf_model.feature_importances_
plt.figure()
plt.barh(X.columns,importance)
plt.xlabel('Importances')
plt.title('Feature Importances - random forest')
plt.show()
```

```python
# confusion Matrix 
from sklearn.metrics import confusion_matrix , ConfusionMatrixDisplay
y_pred = rf_model.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=iris.target_names)
plt.figure()
disp.plot()
plt.title('confusion matrix ')
plt.show()
```
