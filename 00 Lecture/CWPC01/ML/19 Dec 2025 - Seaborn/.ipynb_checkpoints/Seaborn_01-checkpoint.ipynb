```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load sample data 
df = sns.load_dataset("tips")
df.head()
```

```python
sns.set_style("whitegrid") # darkgrid, whitegrid, dark, white, ticks
sns.lineplot(x = "size", y="total_bill",data=df)
plt.title("Total bill vs group size ")
plt.show()
```

```python
# bar plot
sns.barplot(x="day",y="total_bill",data=df)
plt.title("Average total bill by day")
plt.show()
```

```python
# Count plot (Frequency Count)
# number of customer per day
sns.countplot(x="day",data=df)
plt.title("number of customer per day")
plt.show()
```

```python
# histogram (Data distribution)
sns.histplot(df["total_bill"],bins=20,kde=True)
plt.title("Distribution of total bill")
plt.show()
```

```python
# box plot(outlier dectection)
sns.boxplot(x="day",y="total_bill",data=df)
plt.title("total bill distribution by day")
plt.show()
```

```python
# violin plot (Distribution + density)
sns.violinplot(x="day",y="tip",data=df)
plt.title("tips - distribution by day")
plt.show()
```

```python
# scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.title("Total bill vs tip")
plt.show()
```

```python
# regression plot (linear relationship)
sns.regplot(x="total_bill",y="tip",data=df)
plt.title("regression bill vs tip")
plt.show()
```

```python
# heatmap
# create correlation 
corr = df.corr(numeric_only=True)
# plot heatmap

sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("correlation")
plt.show()
```

```python
# pair plot
sns.pairplot(df,hue="sex")
plt.show()
```

```python
# custom color palettes
sns.set_palette("Set2")
# Set1 and Set2
# coolwarm
# pastel
# deep
sns.barplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
fig,axes = plt.subplots(1,2 , figsize=(12,5))
sns.barplot(x="day",y="total_bill",data=df,ax=axes[0])
sns.barplot(x="day",y="total_bill",data=df,ax=axes[1])

plt.show()
```
