```python
# DBSCAN

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
# Dataset 
X = np.array([
    [2,3],[3,4],[2,4],[8,7],[8,8],[25,80]
])
```

```python
# DBSCAN Model 
model = DBSCAN(eps=3,min_samples=2)

labels = model.fit_predict(X)
```

```python
print(labels)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.show()
```
