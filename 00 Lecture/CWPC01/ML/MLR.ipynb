```python
import numpy as np
# step 1
# data set 
X = np.array([
    [1,1000,2],
    [2,1500,3],
    [3,2000,3],
    [4,1200,4],
    [5,1800,2],
])

y = np.array([150,220,280,170,260])

```

```python
# Step 2 fit model using old formula
X_transpose = np.transpose(X)
beta = np.linalg.inv(X_transpose.dot(X)).dot(X_transpose).dot(y)
print('Coefficients (beta) = ',beta)
```

```python
# step 3 make predication for a new house -> size = 1600sq bedrooms = 3
new_house = np.array([1,1600,3])
predicted_price = np.dot(new_house,beta)
print('predicted price for new house = ',predicted_price)
```

```python
# step 4 Evaluate
y_pred = X.dot(beta)
mse = np.mean((y-y_pred)**2)
print("Mean Squared Error : ",mse)
```

```python
# step 4 R - Squared 
ss_total = np.sum((y - np.mean(y))**2)
ss_res = np.sum((y - y_pred)**2)

r_squared = 1 - (ss_res / ss_total)
print(f"r squared = {r_squared}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
```

```python
# step no. 1 prepare data 
data = {
    'size':[1000,1500,2000,1200,1800],
    'Bedrooms':[2,3,3,2,4],
    'Price':[150,220,280,170,260]
}
```

```python
df = pd.DataFrame(data)
X = df[['size','Bedrooms']]
y = df['Price']
```

```python
# step 2 Fir the model
model = LinearRegression()
model.fit(X,y)
print(f'Intercept (B0) = {model.intercept_}')
print(f'Coefficients (B1,B2) = {model.coef_}')
```

```python
# step 3 make predication for a new house -> size = 1600sq bedrooms = 3
new_house1 = np.array([[1600,3]])
predicted_price1 = model.predict(new_house1)
print(f'predicted price for new house = {predicted_price1[0]:.2f}')
```

```python
# step 4 Evaluate
y_pred = model.predict(X)
mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
print(f" Y pred= {y_pred}")
print(f" Mean Squared Y pred= {mse}")
print(f" R2 Squared Y pred= {r2}")
```
