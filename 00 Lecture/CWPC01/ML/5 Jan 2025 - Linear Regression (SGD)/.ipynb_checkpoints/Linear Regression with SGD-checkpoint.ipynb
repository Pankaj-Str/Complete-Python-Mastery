```python
import numpy as np 
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
# sample data 
X = np.array([[1000],[1500],[2000],[2500]]) # features (size* sqft) 
y = np.array([20,30,40,50]) # price in cr
```

```python
# scale features SGD
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# create a model - SGDRegressor
model = SGDRegressor(max_iter=1000,tol=1e-3,learning_rate='constant',eta0=0.01) # eta0 is learning rate
# train the model 
model.fit(X_scaled,y)
```

```python
# predict on new data
new_size = np.array([[1800]]) # new house size 
new_size_scaled = scaler.transform(new_size)

predicted_price = model.predict(new_size_scaled)
print("predicted_price = ",predicted_price[0])
```

```python
# check MSE on traning data
y_pred = model.predict(X_scaled)
mse = mean_squared_error(y,y_pred)
print("mean_squared_error = ",mse)
```
