```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets  import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# load dataset
iris = load_iris()
X = iris.data # features (4 columns - sepal length, sepal width, petal length, petal width)
y = iris.target
features_names = iris.feature_names
print(features_names)
```

```python
print('original shape of the data:', X.shape)
print('features_names of the data:', features_names)
```

```python
# Step 2 - Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# step 3 - Apply PCA
pca = PCA(n_components=2) # reduce to 2 dimensions
X_pca = pca.fit_transform(X_scaled)
print('shape of the data after PCA:', X_pca.shape)
```

```python
# step 4 variance explained by each principal component
explained_variance = pca.explained_variance_ratio_
print('explained variance ratio:', explained_variance)
```

```python
for i ,var in enumerate(explained_variance):
    print(f'Principal Component {i+1} explains {var:.2%} of the variance')
```
