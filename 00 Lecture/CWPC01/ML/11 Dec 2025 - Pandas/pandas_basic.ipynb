```python
# dataset as a dictionary -> (key -> column Name)
data = {
    'Name':['nishant','Prasad','pooja','arya','joy'],
    'Age':[23,12,14,15,20],
    'City':['Mumbai','Surat','Jaipur','Chicago','Houston']
}

df = pd.DataFrame(data)
print(df)
```

```python
# add new column 
df['Salary'] = [20000,34000,450000,67000,870000]
print(df)
```

```python
# assign() 
df = df.assign(Bonus = [300,400,600,578,900])
print(df)
```

```python
# add column with same value for all row
df['NewCountry'] = 'USA'
print(df)
```

```python
# add new column based on calculation
df['total_amounts'] = df['Salary']+df['Bonus']
print(df)
```

```python
# update existing values 
df.loc[2,'Age'] = 32
print(df)
```

```python
# multiple value in one column 
df.loc[2:3,'Salary'] = [75000,85000] 
print(df)
```

```python
# update using condition 
df.loc[df['Salary'] > 50000,'Senior'] = 'Yes'
df.loc[df['Salary'] <= 50000,'Senior'] = 'No'
print(df)
```

```python
# dropping Columns and row
# drop column
df_col_drop = df.drop('City',axis=1)
print(df_col_drop)
```

```python
# multiple drop column 
df_new = df.drop(['Bonus','total_amounts'],axis=1)
print(df_new)
```

```python
# delete row multiple
df_clean = df.drop([0,3])
print(df_clean)
```

```python
# delete single row
df_clean = df.drop(2)
print(df_clean)
```
