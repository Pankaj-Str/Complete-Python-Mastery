```python
import math

# sigmoid turns z into chance (0 to 1)
def sigmoid(z):
    return 1/(1+math.exp(-z))


hours = 2
z = -1.5 + 0.6 * hours
chance = sigmoid(z) # get chance
prediction = "pass" if chance > 0.5 else "fail"
print(f"chance of passing {chance:.2f} ---- {chance*100:.0f}%")
print(f"prediction : {prediction}")
```

```python
# multiple students
hours_list = [2,4,6]
for hours in hours_list:
    z = -1.5 + 0.6 * hours
    chance = sigmoid(z) # get chance
    prediction = "pass" if chance > 0.5 else "fail"
    print(f"chance of passing {chance:.2f} ---- {chance*100:.0f}%")
    print(f"prediction : {prediction}")
    
```

```python
# maximum likelihood Estimation (MLE)

from sklearn.linear_model import LogisticRegression
import numpy as np

# data (hour need )
hours = np.array([[2],[4],[6]])
answer = [0,1,1] # fail = 0 , pass = 1

# make model 
model = LogisticRegression()
model.fit(hours,answer) # MLE

print("Starting Point = (a0) : ",model.intercept_[0])
print("chance per hour (a1) :",model.coef_[0][0])

# make guesses
guesses = model.predict(hours)
print("guesses :",guesses)
```
