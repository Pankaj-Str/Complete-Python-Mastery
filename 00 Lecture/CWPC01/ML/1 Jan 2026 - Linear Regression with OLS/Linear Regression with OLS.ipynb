```python
import numpy as np
```

```python
# Data
X = np.array([800,1000,1200,1500,1800]) #house size 
y = np.array([180,220,260,300,350]) #prices
```

```python
# calculate means
X_mean = np.mean(X)
y_mean = np.mean(y)
```

```python
# calculate slope(w) and intercept (b)
w = np.sum((X - X_mean) * (y - y_mean))/ np.sum((X - X_mean)**2)
b = y_mean - w * X_mean
```

```python
print(f"Slope (w) {w:6f}")
print(f"Intercept (b) {b:6f}")
```

```python
predictions = w * X + b
print("predictions",np.round(predictions,2))
```

```python
from sklearn.linear_model import LinearRegression

X = np.array([800,1000,1200,1500,1800]).reshape(-1,1) #house size 
y = np.array([180,220,260,300,350]) #prices
```

```python
# create and fit model 
model = LinearRegression()
model.fit(X,y)
```

```python
print(f"Slope (w) {model.coef_[0]:.6f}")
print(f"Intercept (b) {model.intercept_:6f}")
```

```python
predictions = model.predict(X)
print("predictions",np.round(predictions,2))
```

```python
# Predict a new house (1300 Sq..)
new_size = np.array([[1300]])
predicted_price = model.predict(new_size)
print(f"predicted price for 1300 = price {predicted_price[0]:.2f}k")
```
