```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# dataset 
data = [[40],[34],[50],[60],[70],[85],[90],[100],[25]]
```

```python
model = KMeans(n_clusters=2)
model.fit(data)
```

```python
plt.scatter(data,[1,2,3,4,5,6,7,8,9],c=model.labels_)
plt.show()
```
