```python
import matplotlib.pyplot as plt
```

```python
# line plot -> growth , decline , trend over time  , sales over month..etc

days = [1,2,3,4,5] # x axis data
temperature = [22,24,19,23,25] # y axis data

plt.plot(days,temperature)
plt.xlabel("Days")
plt.ylabel("temperature")
plt.title("Daily temperature data")
plt.show()
```

```python
# bar chart (category comparison)

students = ["joy","prasad","nishant","neelam","heena","rahul"]
marks = [34,85,78,69,54,20]
plt.bar(students,marks)
plt.xlabel("Student Name")
plt.ylabel("Student Marks")
plt.title("Student Marks Comparsion")
plt.show()
```

```python
# move to horizontal bar chart 

plt.barh(students,marks)
plt.xlabel("Student Name")
plt.ylabel("Student Marks")
plt.title("Student Marks Comparsion")
plt.show()
```

```python
# scatter plot (relationship analysis)

study_hours = [1,2,3,4,5,6]
marks = [34,85,78,69,54,20]

plt.scatter(study_hours,marks)
plt.xlabel("Study Hour")
plt.ylabel("Marks")
plt.title("Study hour vs Marks")
plt.show()
```

```python
# histogram (data distribution)

marks = [12,13,14,15,16,20,30,40,50,60,70,77,88,45,90]

plt.figure(figsize=(8,5))
plt.hist(
    marks,
    bins=5,
    edgecolor="black", # border color
    alpha = 0.8 # transparency
)
plt.xlabel("marks",fontsize = 15)
plt.ylabel("Frequency",fontsize = 15)
plt.title("Marks Distrubution")
plt.grid(axis="y",linestyle='--',alpha = 0.2)
plt.tight_layout()
plt.show()
```

```python
# pie chart (percentage share)

students = ["joy","prasad","nishant","neelam","heena","rahul"]
marks = [34,85,78,69,54,20]

plt.pie(marks,labels=students,autopct="%1.1f%%")
plt.title("Programming % data ")
plt.show()
```

```python
# multiple line plot(comparision)
month = [1,2,3,4,5]
sales_2024 = [200,300,230,400,100]
sales_2025 = [300,400,500,200,50]

plt.plot(month,sales_2024 , label="2024 Sales")
plt.plot(month,sales_2025 , label="2025 Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("yearly sales ")
plt.legend()
plt.show()
```

```python
# box plot (outliers detection)
salary = [20000,45000,67000,30000,32000,56000]
plt.boxplot(salary)
plt.show()
```

```python
# area plot (cumulative data)

years = [2000,2001,2002,2003,2004]
profit = [10,20,35,45,20]
plt.fill_between(years,profit,alpha=0.4)
plt.plot(years,profit,marker="o",linewidth=2,label="profit")
plt.xlabel("year")
plt.ylabel("profit")
plt.title("Comany Profit Growth")
plt.xticks(years)
plt.grid(True,linestyle="--",alpha=0.6)
plt.legend()
plt.show()
```
