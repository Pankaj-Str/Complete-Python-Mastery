```python
# intertools -> fast tool for loop
import itertools

for i in itertools.count(start=1,step=2):
    # stop 
    if i > 9:
        break
    print(i)
```

```python
# cycle() - > repeat a sequence
color = ["red","green","blue"]
count = 0
for c in itertools.cycle(color):
    print(c)
    count += 1
    if count == 7:
     break
```

```python
# repeat() -> repeat value
for x in itertools.repeat("welcome",3):
    print(x)
```

```python
# accumulate() -> runing total
nums = [1,2,3,4,5,6]
result = itertools.accumulate(nums)
print(list(result))
```

```python
# combinations() - > choose item without orders
items = ["A","B","C","d","e","f"]
for combo in itertools.combinations(items,2):
   print(combo)
```

```python
# product() -> cartesan product

size = ["s","m"]
color = ["red","blue"]

for p in itertools.product(size,color):
    print(p)
```

```python
# islice -> slice an iterator like a list
nums = range(100)
first_five = itertools.islice(nums,0,5) # 0 to 4
print(list(first_five))
```
