```python
import functools
import time
```

```python
# lru_cache -> speed up repeated function 
def slow_function(n):
    time.sleep(2) # similate heavy work
    return n * n

print(slow_function(5))
print(slow_function(50))
print(slow_function(100))
```

```python
@functools.lru_cache(maxsize=None)
def fast_function(n):
    time.sleep(2) # similate heavy work
    return n * n

print(fast_function(5)) # slow first time
print(fast_function(15))
print(fast_function(100))
```

```python
# show chash statistics
print(fast_function.cache_info())
```

```python
# reduce() -> reduce a list to a single value

number = [1,2,3,4,5]
result = functools.reduce(lambda a,b : a+b,number)
print(result)

result = functools.reduce(lambda a,b : a*b,number)
print(result)
```

```python
# cmp_to_key()
def compare(a,b):
    if len(a) < len(b):
        return -1
    elif len(a) > len(b):
        return 1
    else: 
        return 0
    
words = ['apple','kiwi','banana','mango']
sorted_word = sorted(words,key=functools.cmp_to_key(compare))
print(sorted_word)  

words.sort()
print(words)  
```
