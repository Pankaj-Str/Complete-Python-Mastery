```python
products = []
gst_rate = 0
discount_details = {"type":None, "value":0,"product_id":None}

def add_product(): # add product
    try:
        pid = input("Enter Product ID : ")
        name = input("Enter Product Name")
        qty = int(input("Enter Quantity : "))
        price = float(input("Enter Unit Price "))
        
        products.append(
            {
                "id":pid,
                "name":name,
                "qty":qty,
                "price":price
            }
        )
        print("product add successfully \n")
        print(products)
    except ValueError:
        print("Invalid Input ! ... Quantity and price must be number....")    
              
```

```python
def view_product():
    if not products:
        print("No Products added yet \n")
        return
    print("Product list ------------")
    print("Product ID | Product Name | Quantity | Unit Price")
    for p in products:
        print(f"{p['id']}   |   {p['name']}   |.   {p['qty']}   |.  {p['price']}")
    print("------------------------------------\n")    
```

```python
# remove product 
def remove_product():
    if not products:
        print("No Product Remove ..\n")
        return
    pid = input("Enter Proudct Id to remove item from list")
    for p in products:
        if p['id'] == pid:
            products.remove(p)
            print("Product Remove Successfully ....")
            return
    print("Product ID not Found....")    
```

```python
# apply gst
def apply_gst():
    global gst_rate
    try:
        gst_rate = float(input('Enter GST rate % '))
        print(f"GST @ {gst_rate}% apply successfully....")
    except ValueError:
        print("invaild GST rate... \n")
            
```

```python
# Discount Offers
def apply_discount():
    global discount_details
    print("Select Discount Type ")
    print("1 Percentage Discount on the total bill ")
    print("2 Fixed Amount Discount on the total bill ")
    print("3 Discount on specific products ")
    
    try:
        choice = int(input("Enter Choice "))
        if choice == 1:
            rate = float(input("Enter Discount Rate % ..."))
            discount_details = {"type":"Percentage","value":rate}
        elif choice == 2:
            amount = float(input("Fixed Amount Discount Amount /- ..."))
            discount_details = {"type":"Fixed","value":amount}     
        elif choice == 3:
            pid = input("Enter Product ID For Discount ...")
            rate = float(input("Enter Discount Rate % ..."))
            discount_details = {"type":"product","value":rate,"product_id":pid} 
        else:
            print("Invalid Choice..\n")
            return
        print("Discount applied Successfully....")
    except ValueError:
        print("Invaild Input \n")               
```

```python
add_product()
view_product()
# remove_product()
apply_gst()
```
