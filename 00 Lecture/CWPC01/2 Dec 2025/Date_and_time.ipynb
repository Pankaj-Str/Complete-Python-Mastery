```python
import datetime

# get current date and time 
now = datetime.datetime.now()
print(now)

# current time only 
time_now = datetime.datetime.now().time()
print("Current Time = ",time_now)
```

```python
# set custom date 

d = datetime.date(2025,12,2)
print(d)
# custom time 
t = datetime.time(10,30,45)
print(t)
```

```python
# format date and time 

now = datetime.datetime.now()
print(now)

print(now.strftime("%Y-%m-%d"))
print(now.strftime("%d-%m-%y"))
print(now.strftime("%H:%M:%S"))
print(now.strftime("%A")) # Day names
print(now.strftime("%B")) # Month name
print(now.strftime("%I")) # Hours
print(now.strftime("%p")) # AM/PM
```

```python
# convert String to date(strptime)
date_str = "20-05-2025"

print(type(date_str))
print(date_str)

d = datetime.datetime.strptime(date_str,"%d-%m-%Y")
print(d)
print(type(d))
```

```python
# add , subtract Days ,Months , Time

today =  datetime.date.today()
print(today)

tomorrow = today + datetime.timedelta(days=1)
print("tomorrow = ",tomorrow)

yesterday = today - datetime.timedelta(days=1)
print("yesterday = ",yesterday)

# add hour / minutes
now = datetime.datetime.now()
after_5_hours = now + datetime.timedelta(hours=5)
print("after 5 hours = ",after_5_hours)
```

```python
# get timestamp
import time
timestamp = time.time()
print(timestamp)
```

```python
# sleep / Delay Program 
print("Wait 3 seconds...")
time.sleep(3)
print("done..!")
```

```python
# get UTC Time
utc = datetime.datetime.utcnow()
print("utc time = ",utc)
```

```python
# difference between two dates 

from datetime import date

d1 = date(1990,5,20)
d2 = date(2025,12,2)
difference = d2- d1
print(difference)
```

```python
from datetime import datetime

dt1 = datetime(2024,5,20,10,30,0)
dt2 = datetime(2025,12,2,12,45,30)

difference = dt2 - dt1
print(difference)
print("Days = ",difference.days)
print("seconds = ",difference.seconds)
print("total seconds = ",difference.total_seconds())

hours = difference.total_seconds()/3600
minutes = difference.total_seconds()/60

print("Hours = ",hours)
print("Minutes = ",minutes)
```
