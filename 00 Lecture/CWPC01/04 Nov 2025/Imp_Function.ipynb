```python
# function 
# enumerate -> add index to each item in an iterable
# when we use - > need to both position and value 

# enumerate(iterable, start=0)

# without enumerate
seasons = ['Spring', 'Summer', 'Fall', 'Winter']

print("Without enumerate:")

for i in range(len(seasons)):
    print(i, seasons[i])  
```

```python
i = 101
for s in seasons:
    print(f'{i} : {s}')
    i += 1
```

```python
# using enumerate function
# enumerate(interable, start=0)

for i , season in enumerate(seasons ,start=101):
    print(i, season)
```

```python
# 2nd function
# Zip function
# zip(*iterables) -> combine multiple iterables into tuples
# use -> when you want to loop over two or more list "in parallel"
```

```python
name = ['Pankaj', 'Ankit', 'Rohit']
age = [24, 25, 26]

for i in range(len(name)):
    print(f'{name[i]} is {age[i]} years old.')
```

```python
# using zip function
for Name , Age in zip(name, age):
    print(f'{Name} is {Age} years old.')
```

```python
# extra 
extra = [23, 45 ]

for Name , Age, Ex in zip(name, age, extra):
    print(f'{Name} is {Age} years old and extra is {Ex}.')
```

```python
# rohit and 26 are ignored because extra has only two elements

# unpacking with zip
pairad  = list(zip(name, age))
print(pairad)

name_back, age_back = zip(*pairad)
print(name_back)
print(age_back)

# converting back to list
name_list = list(name_back)
age_list = list(age_back)
print(name_list)
print(age_list)
```

```python
# any funciton -> retun true if at least one element is truthy
# use -> quick " is there anything true ?" check 

# return true if any element is true (non-zero, non-empty,True,etc)

numbers = [0, 0, 5, 0]
print(any(numbers))  # True because 5 is non-zero

flags = [False, False, True]
print(any(flags))    # False because all are False
```

```python
#  empty interable
empty_list = []
print(any(empty_list))  # False because there are no elements
```

```python
# checking use input 
answer = ["yes", "y", "no",'']
has_yes = any(answer.lower().startswith('y') for answer in answer)
print(has_yes)  # True because "yes" and "y" start with 'y
```

```python
# all function -> return true id  every element is truthy
# use -> Quick " are all conditions met ?" check

positive_numbers = [1, 2, 3, 4]
print(all(positive_numbers))  # True because all are non-zero

mixed_numbers = [1, 0, 3, 4]
print(all(mixed_numbers))     # False because 0 is falsy

# empty iterable
empty_list = []
print(all(empty_list))        # True because there are no falsy elements
```
