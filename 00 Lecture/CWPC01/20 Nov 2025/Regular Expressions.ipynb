```python
# Regular Expressions in Python
import re

# re.match()
# re.search()
# re.findall()
# re.sub()
```

```python
# re.match() -> matches only at the beginning of string 

text = "codes with pankaj"

match = re.match(r"code",text)

if match:
    print("Found : ",match.group())
    print("Position : ",match.span())
else:
    print("No Match Found")    
```

```python
# re.search() -> finds the first match anywhere is the string
text = "the rain is spain stays mainly in the plain spain"

match = re.search("spain",text)

if match:
    print("Found : ",match.group())
    print("Position : ",match.span())
else:
    print("No Match Found")
     
```

```python
# re.findall() -> return all matches as a list 

match = re.findall("spain",text)
print(match)
```

```python
text = "my number are 123,456 and 340"
numbers = re.findall(r"\d+",text)
print(numbers)
```

```python
# find the only text start with s
text = "welcome to codes with pankaj and codes with best to codes"
words = re.findall(r"c\w+",text)
print(words)
```

```python
# re.sub() -> replace matches with new text
new_text = re.sub(r"codes","java",text)
print(new_text)
```
