```python
from collections import Counter

fruits = ['apple','banana','apple','orange','apple','banana']

count = Counter(fruits)
print(count)
```

```python
# most common items 
text = "banana"
c = Counter(text)
print(c.most_common(2)) # top 2 characters
```

```python
# default dict (dictionary with default values)
# data = {}
# print(data["count"])

from collections import defaultdict

data = defaultdict(int)
data["apple"] += 1
data["apple"] += 1
data["banana"] += 1

print(data)
```

```python
# grouping items 
from collections import defaultdict


students = [
    ("amit","python"),
    ("Neha" ,"java"),
    ("ravi","python"),
    ("prasad","c++")
]

group = defaultdict(list)
print(group)

for name , course in students:
    group[course].append(name)
    
print(group)    
```
