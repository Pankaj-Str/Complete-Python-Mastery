```python
import math

print("PI = ",math.pi)
print("Euler's Number e ",math.e)
```

```python
number = 16
result = math.sqrt(number)
print("Square root = ",result)
```

```python
# power and absolute value 
print(math.pow(2,3))
print(abs(-5))
print(math.fabs(-5))
```

```python
# rounding  functions 
x = 3.7
print("floor ",math.floor(x))
print("ceil : ",math.ceil(x))
```

```python
# trigonometrics function 

angle_deg = 30
angle_rad = math.radians(angle_deg)
print("radians : ",angle_rad)
print(' sin(30) = ',math.sin(angle_rad))
print(' cos(30) = ',math.cos(angle_rad))
print(' tan(30) = ',math.tan(angle_rad))
```

```python
# random module 
import random as r

for i in range(3):
    print(r.random())
```

```python
# random int 

dice = r.randint(1,10)
print("Dice value = ",dice)
```

```python
# choice 
color = ["red","green","blue","yellow"]
print(r.choice(color))
```

```python
cards = ["A","K","T","U","Z"]
r.shuffle(cards)
print(cards)
```

```python
# OTP (4 digit)
otp = ""
for i  in range(4):
    otp += str(r.randint(0,9))
    
print("your OTP is ",otp)    
```
