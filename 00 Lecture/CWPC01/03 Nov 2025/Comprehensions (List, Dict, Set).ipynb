```python
# List Comprehension
# list : [expression for item in iterable if condition]
# dict Comprehension
# dict : {key_expression: value_expression for item in iterable if condition}
# set Comprehension
# set : {expression for item in iterable if condition}
```

```python
# list - Basic
# Squares of numbers 0-9
squares = [x**2 for x in range(10)]
# print(type(squares)) 
print(squares)
```

```python
data = []
for i in range(10):
    i = i**2
    data.append(i)
    
print(data)    
```

```python
# with condition 
# even numbers 

evens = [x for x in range(20) if x%2 == 0 ]
print(evens)
```

```python
# nexted loops

matrix = [[1,2,3],[4,5,6],[7,8,9]]
print(matrix)

flat = [num for row in matrix for num in row]
print(flat)
```

```python
# 2 Dictionary Comprehension

data = {"name":"Nishant","age":23}
print(type(data))
print(data)

print(data['name'])
```

```python
# {number:square}
# number = x
square_dict = {x: x**2 for x in range(5)}
print(square_dict)

print(square_dict[3])
```

```python
names = ['Rahul',"prasad","Nishant","Tiya"]

# name:length
# len(names)

# names1 = names[0]

# size = len(names1)
# print(size,names[0])


for i in names:
    print(f'{len(i)} = {i}')
    
    
lengths = {i:len(i) for i in names}
print(lengths)    
```

```python
# inverting a dictionary

d = {'a':1,'b':2,'c':3}

# v = 1
# k = a
inverted = {v:k for k , v in d.items()}
print(inverted)
```

```python
num1 , num2 , num3 = 80,90,70
print(num1)
print(num2)
print(num3)

num3 , num2 , num1 = num1 , num2 , num3

print(num1)
print(num2)
print(num3)
```

```python
# set  Comprehension

my_data = [12,12,13,14,1,1,1,1,1]

data = set(my_data)
print(type(data))
print(data)
```

```python
my_data = [12,12,13,14,1,1,1,1,1]
# unique squares
unique_squares = {x**2 for x in my_data}
print(unique_squares)
```

```python
# unique letters in word

word = "codes with pankaj chouhan"
 
letters = {char.lower() for char in word}
print(letters)
```
