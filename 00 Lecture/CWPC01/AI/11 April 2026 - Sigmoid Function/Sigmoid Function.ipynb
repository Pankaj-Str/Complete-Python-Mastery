```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
z = np.linspace(-10,10,100) # 100 points generate values -10 to 10
```

```python
sig = sigmoid(z)
sig
```

```python
# a = sigmoid(23)
# a
```

```python
# plot 
plt.plot(z,sig)
plt.xlabel("Input Z")
plt.ylabel('sigmoid(z)')
plt.grid(True)
plt.axhline(0.5 , color='r',linestyle='---')
plt.show()
```
