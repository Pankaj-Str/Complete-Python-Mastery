```python
import threading as th
import time

def task1():
    print("start task 1")
    time.sleep(2) # simulate work
    print("finished task 1")
    
    
def task2():
    print("start task 2")
    time.sleep(1) # simulate work
    print("finished task 2")
    
# create threads
t1 = th.Thread(target=task1)
t2 = th.Thread(target=task2)

# start threading
t1.start()
t2.start()

# wait for both to finish
t1.join()
t2.join()

print("all task done ! ")
```

```python
# synchronization
counter = 0 

lock = th.Lock()
def increment():
    global counter
    for _ in range(300000):
        with lock:
          counter += 1
        
t11 = th.Thread(target=increment)
t12 = th.Thread(target=increment)
t11.start()
t12.start()
t11.join()
t12.join()
print(f"Final counter : {counter}")
        
```

```python
import queue
import threading as th
import time


def producer(q):
    for i in range(5):
        time.sleep(1) # simulate time taken to produce an item
        q.put(i)
        print(f"Produced {i}")
        
        
def consumer(q):
    while True:
        item = q.get()
        if item is None: # stop signal
            break
        print(f"Consumed {item}")
        q.task_done()
        
        
q = queue.Queue()
thread_prod = th.Thread(target=producer, args=(q,))
thread_cons = th.Thread(target=consumer, args=(q,))

thread_cons.start()
thread_prod.start()


thread_prod.join()
q.put(None) # send stop signal to consumer
thread_cons.join()            
```
