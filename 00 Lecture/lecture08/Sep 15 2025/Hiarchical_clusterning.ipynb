```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram , linkage
import matplotlib.pyplot as plt
```

```python
data = np.array([[1],[5],[8],[10],[19],[20]])
# compute linkage matrix
Z = linkage(data,method='single')

plt.Figure(figsize=(4,6))
dendrogram(Z,labels=['1','5','8','10','19','20'])
plt.title('hierarchical clustering')
plt.xlabel('Data Point')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
data = np.array([
    [10,5], # customer 1 
    [12,7], # customer 2 
    [20,15], # customer 3 
    [22,18], # customer 4 
    [50,40], # customer 5 
    [55,45], # customer 6 
])
Z = linkage(data,method='single', metric='euclidean')
```

```python
# Agglomerative Clustering 
# set number clusters to 3 for demosstration

agg_clustering = AgglomerativeClustering(n_clusters=3,linkage='single',metric='euclidean')
plt.Figure(figsize=(10,5))
dendrogram(Z,labels=['C1','C2','C3','C4','C5','C6'])
plt.title('dendrogram for single linkage')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# -------
plt.Figure(figsize=(10,5))
scatter = plt.scatter(data[:, 0],data[:, 1],cmap='virdis',s=100)
plt.title('Customers Clusters')
plt.xlabel('Annual Spend ')
plt.ylabel('Purchase')
plt.legend(*scatter.legend_elements(),title='clusters')
plt.grid()
plt.show()
```
