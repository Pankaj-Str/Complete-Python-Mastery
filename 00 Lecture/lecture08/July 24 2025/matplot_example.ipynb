```python
import matplotlib.pyplot as plt
import numpy as np
# data 
x = np.array([1,2,3,4,5])
y = np.array([2,3,5,9,10])

# create a line plot
plt.plot(x,y)
plt.show()
```

```python
# create a scatter plot
plt.scatter(x,y)
plt.show()
```

```python
# data 
cat = ["a","b","c","d","e"]
value = [4,3,10,5,2]

# create a bar

plt.bar(cat,value)
plt.show()
```

```python
# create a histogram 
data = np.random.randn(1000)

plt.hist(data,bins=30)
plt.show()
```

```python
data = [np.random.randn(100),np.random.randn(100)+2,np.random.randn(100)-2]
# create a box plot(data)
plt.boxplot(data)
plt.show()
```
