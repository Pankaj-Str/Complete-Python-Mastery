```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DataSet
x = np.array([24, 50, 5, 9, 190, 100,56]).reshape(-1,1) # Temperatures
y = np.array([3, 8, 98, 180, 120, 22,0]) #sales 
```

```python
# model 
model = LinearRegression()
model.fit(x,y)
```

```python
m = model.coef_[0] # Slope (m)
b = model.intercept_

print(f'slope = (m) {m:.2f}')
print(f'intercept = (b) {b:.2f}')
```

```python
# Prediction 75
Prediction = model.predict([[5]])[0]
print(f'Prediction  sales {Prediction:.2f}')
```

```python
plt.scatter(x,y,color="blue",label='Data Points')
plt.plot(x,model.predict(x),color="red",label="Best Fir Line")
plt.xlabel('Temperature (F)')
plt.ylabel('Ice Cream Sales')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
