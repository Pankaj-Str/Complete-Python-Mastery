```python
def linear_regression_ols(x,y):
    n = len(x) # number of data points
    if n != len(y):
        raise ValueError("x and y must have the same length")
    
    # calculate means
    x_mean = sum(x) / n
    y_mean = sum(y) / n
    
    
    # calculate the slope (b1)
    
    numerator = sum((xi - x_mean) * (yi - y_mean) for xi, yi in zip(x, y))
    denominator = sum((xi - x_mean) ** 2 for xi in x)
    
    # slope (m)
    
    m = numerator / denominator if denominator != 0 else 0
    
    #intercept (b)
    b = y_mean - m * x_mean
    
    return m, b
    
```

```python
# use input data 

x = [24,50,50,90,90,100] # temperature in Fahrenheit
y = [3,8,7,18,20,22] # ice cream sales in thousands

m , b = linear_regression_ols(x, y)
print(f"Slope (m): {m:2f}")
print(f"Intercept (b): {b:2f}") 
```

```python
# make predictions

predictions = m * 85 + b 
print(f"Predicted sales at 75F: {predictions:2f} thousands")
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(x, y, color='blue', label='Data points')

x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', label='Regression line')

plt.xlabel('Temperature (F)')
plt.ylabel('Ice Cream Sales (thousands)')
plt.title('Linear Regression OLS')
plt.legend()
plt.grid()
plt.show()
```
