```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plots 
# create data 
x = np.array([1,2,3,4,5])
y1 = np.array([2,4,6,8,10])
y2 = np.array([1,3,5,7,9])

# create a line plot 
plt.plot(x,y1, label="dataset 1")
plt.plot(x,y2, label="dataset 2")

# add title 
plt.title("Customized plots")
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')

# add legend
plt.legend()
plt.show()
 
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])


plt.scatter(x,y, color="red",marker='*',s=100)
# add title 
plt.title("Customized plots")
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# create a line 
plt.plot(x,y)

# set axis limits
plt.xlim(0,6)
plt.ylim(0,12)

# customize ticks
plt.xticks(np.arange(0,7,1))
plt.yticks(np.arange(0,13,4))
plt.title("Customized plots")
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
# create a scatter plot
plt.scatter(x,y)
plt.annotate('Highest Point',xy=(5,10),xytext=(4,8),
             arrowprops=dict(facecolor='black',arrowstyle='->'))
# customize ticks
plt.xticks(np.arange(0,7,1))
plt.yticks(np.arange(0,13,4))
plt.title("Customized plots")
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
