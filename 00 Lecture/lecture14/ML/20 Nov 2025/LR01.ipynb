```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# data 
X = np.array([8,10,19]).reshape(-1,1) # pizza Diameters (independent variable)
Y = np.array([10,13,16]) # pizza price (dependent variable)
```

```python
# create and train the model
# Y = mX + b
model = LinearRegression()
model.fit(X,Y)
```

```python
# get the slope (m) 
# get the Intercept (b)
m = model.coef_[0] # Slope 
b = model.intercept_ # Intercept

print(f"slope : {m:.1f}")
print(f"Intercept : {b:.2f}")
```

```python
# predict price for 20inch = price ?
x_new = np.array([20]).reshape(-1,1)
y_predict = model.predict(x_new)
print(f"Predicted price for 20 inch pizza = {y_predict[0]:.2f}")
```

```python
# plot the data point 
plt.scatter(X,Y, color='blue',label="data point")
plt.plot(X,model.predict(X),color="red",label="Regression Line")
plt.xlabel("Diameter (inch)")
plt.ylabel("Price $")
plt.title("Pizza Price Prediction")
plt.legend()
plt.show()
```
