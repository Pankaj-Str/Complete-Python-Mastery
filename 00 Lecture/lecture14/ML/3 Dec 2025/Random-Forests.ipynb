```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
# dataset age and income
X = np.array([
    [25,30000],
    [30,50000],
    [20,20000],
    [35,60000],
    [28,45000]
])
y = np.array([0,1,0,1,1])

# 3 trees with bagging
trees = []
for i in range(3):
    # bootstrap sample
    indices = np.random.choice(len(X),size=len(X),replace=True)
    X_sample = X[indices]
    Y_sample = y[indices]
    
    # train the trees
    tree = DecisionTreeClassifier()
    tree.fit(X_sample,Y_sample)
    trees.append(tree)
    
# predict for a new person age = 27 income = 40000
new_person = np.array([[27,40000]])
vote = [tree.predict(new_person)[0] for tree in trees]
print("tree vote = ",vote) 

final_guess = max(set(vote),key=vote.count) # majorty vote
print("bagging says = ","buy" if final_guess == 1 else "No Buy")
  
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
# print(X)
y = iris.target
# print(y)

# split data 
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.5,random_state=42)

model = RandomForestClassifier(n_estimators=300,random_state=42)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```
