```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
customer = np.array([
    [20,500], #c1
    [40,1000], #c2
    [30,800], #c3
    [18,3000], #c4
    [25,100], #c5
    [100,50] #c6
])
```

```python
kmeans = KMeans(n_clusters=2,random_state=0)
kmeans.fit(customer)

groups = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
plt.scatter(customer[:, 0],customer[:, 1], c=groups,s=50,cmap='viridis',label='customer')
plt.scatter(centers[:, 0],centers[:, 1], c="red",s=100,marker='X',label='center')
plt.title("Kmeans Clusters")
plt.xlabel("Shop per month")
plt.ylabel("spending (INR)")
plt.legend()
plt.show()
```

```python
# Elbow Trick to pick the best K
wcss = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(customer)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 6), wcss, marker='o')
plt.title('Elbow Trick to Pick K')
plt.xlabel('Number of Groups (K)')
plt.ylabel('Spread (WCSS)')
plt.show()
```
