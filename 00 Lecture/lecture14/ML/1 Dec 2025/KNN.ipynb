```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# weight AND price
x = np.array([
    [201,150], # samsung galaxy
    [206,999], # iphone
    [188,200], # one plus
    [197,599]  # pixel
])

# label categories ( 0 = cheap, 1 = expensive)
y = np.array([0, 1, 0, 1])

# create a training model
model = KNeighborsClassifier(n_neighbors=3) # k=3
model.fit(x, y)

# predict one plus with weight 184g and price $300
prediction = model.predict([[184, 300]])
print(prediction)  # Output: [0] (cheap)

# interpret the prediction
category = "expensive" if prediction[0] == 1 else "cheap"
print(f"The predicted category for the phone is: {category}")

```
