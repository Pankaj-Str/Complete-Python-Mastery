```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
```

```python
data = {
    'Size': [1000,1500,2000,2200,2800],
    'Bedrooms': [2,3,4,5,6],
    'Price':[150,220,280,285,290]
}
```

```python
df = pd.DataFrame(data)
X = df[['Size','Bedrooms']]
y = df['Price']
```

```python
# step 2 fit the model
model = LinearRegression()
model.fit(X,y)
print("Intercept:(B0)", model.intercept_)
print("Coefficients: (B1,B2)", model.coef_)
```

```python
# predict the values 1600 sqft and 3 bedrooms
new_house = np.array([[1600,3]])
predicted_price = model.predict(new_house)
print(f"Predicted price for 1600 sqft and 3 bedrooms: {predicted_price[0]:.2f}")
```

```python
# step 4 evaluate the model
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
r2 = r2_score(y, y_pred)
print(f"R-squared: {r2:.2f}")
```
