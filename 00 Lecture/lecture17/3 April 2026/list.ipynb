```python
data = [12,13,14,15,166,17,18,90]
print(data)
```

```python
# access data from list
print(data[4])
print(data[-3])
```

```python
print(data[2:5])
print(data[2:])
print(data[:5])
print(data[:])
```

```python
print(data[-4:-2])
```

```python
# List Methods
# append()
data.append(200)
print(data)
```

```python
# extend()
add = [22,33,44,55]
data.extend(add)
data
```

```python
# insert()
data.insert(4,999)
data
```

```python
# remove()
data.remove(999)
data
```

```python
# pop()
remove_value = data.pop()
print(remove_value)
print(data)
```

```python
# index()
print(data.index(17))

```

```python
data = [2,2,2,2,2,4,5,6,7,8,8,7,6]
# count()
print(data.count(2))
```

```python
data = [12, 13, 14, 166, 17, 8, 9, 2, 22]

# sort()
data.sort()
data
```

```python
# reverse()
data.reverse()
data
```

```python
# copy()
a = data.copy()
a
```

```python
a = data
a
```

```python
# clear()
data.clear()
```

```python
# change and update 
data = [166, 22, 17, 14, 13, 12, 9, 8, 2]
data[2] = 100
print(data)
```

```python
# using loop 
for i in data:
    print(i)
```
