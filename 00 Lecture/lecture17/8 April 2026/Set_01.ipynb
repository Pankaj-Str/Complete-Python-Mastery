```python
# create a set -
data = {21,22,33,22,21,21,21,34,34,32}
print(data)
```

```python
# list data 
list_data = [21,22,33,22,21,21,21,34,34,32]
# set()
data = set(list_data)
print(data)

list_data = list(data)
list_data
print(type(list_data))
```

```python
# add()
data.add(99)
data
```

```python
# update() 
my_list = [11,11,12,14,15,15]

data.update(my_list)
data
```

```python
# remove()	Removes item	Yes (raises error)
data.remove(21)
data
```

```python
# discard()	Removes item (safer)	No
data.discard(99)
data
```

```python
# pop()	Removes and returns one random item	No (works on empty too)
data.pop()
data
```

```python
# clear()	Removes all items	No
data.clear()
data
```

```python
fruits = {"apple", "banana", "cherry"}

print("apple" in fruits)     # True
print("mango" in fruits)     # False
```

```python
for i in fruits:
    print(i)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)      # Union: {1,2,3,4,5,6}
print(A & B)      # Intersection: {3,4}
print(A - B)      # Difference: {1,2}
print(A ^ B)      # Symmetric difference: {1,2,5,6}
```

```python
# Create a set of squares
squares = {x**2 for x in range(1, 6)}
print(squares)          # {1, 4, 9, 16, 25}
```

```python
# Set of even numbers from a list
numbers = [1, 2, 3, 4, 5, 6]
evens = {x for x in numbers if x % 2 == 0}
print(evens)   
```
