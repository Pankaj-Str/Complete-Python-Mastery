```python
df = pd.read_csv("https://raw.githubusercontent.com/Pankaj-Str/Complete-Python-Mastery/refs/heads/main/53%20DataSet/credit_card_transactions.csv")
df.head()
```

```python
# check size 
print(df.shape)
```

```python
# print only columns name - 
df.columns.to_list()
```

```python
# summay Statitistics
df.describe(include='all')
```

```python
df.isnull().sum()
df.dtypes
```

```python
df['Merchant'] = df['Merchant'].fillna('Unknown')
df.isnull().sum()
```

```python
# convert into str to int/float columns 
df['Amount'] = df['Amount'].astype(float)
df.dtypes
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median)
df.isnull().sum()
```

```python
# Convert Transaction_Date into datetime format and extract year, month, and day as new columns.
print(df.head()) # print 5 rows 
print(df.dtypes) # check datatype

# df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'],dayfirst=True)
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])

print(df.dtypes) # check datatype
```

```python
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day

df.head()
# check info
df.info()
```

```python
# Retrieve all transactions made in January 2024.
jan_2024 = df[(df['Year'] == 2023) & (df['Month'] == 2)]
jan_2024.head()
# create a report 
jan_2024.to_csv("feb_2023.csv")
```

```python
# Find transactions where Amount > 1000 and Transaction_Type is "Online".
high_online_t =  df[(df['Amount'] > 1000) & (df['Transaction_Type'] == 'Online') ]
print(high_online_t.head())
# Select only Approved transactions from the dataset.
```
