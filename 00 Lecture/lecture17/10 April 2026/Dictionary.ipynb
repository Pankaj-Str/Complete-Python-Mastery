```python
# Dictionary

data = {
    "name" : "Payal",
    "age": 20,
    "height":5.6
}
```

```python
print(data)
```

```python
# Accessing Values
data["height"]
```

```python
#  Modifying and Adding Items
# update 
data["height"] = 5.9
data
```

```python
# Adding Items
data['city'] = 'Mumbai'
data
```

```python
# Checking Key Existence
# print('name' in data)

if "city" in data:
    print("city is found...",data['city'])
else:
    print("city not found...")
```
