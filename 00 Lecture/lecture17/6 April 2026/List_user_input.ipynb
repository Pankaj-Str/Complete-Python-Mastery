```python
# -- User input section
# Enter Product List Size : 4
# Enter Product 1 -
# Samosa
# Enter Product 2 -
# Kachori
# Enter Product 3 -
# Fafda
# Enter Product 4 -
# Jalebi

# Enter Samosa Price :
# 300/-
# Enter Kachori Price :
# 100/-
# Enter Fafda Price :
# 100/-
# Enter Jalebi Price :
# 200/-

# --- output

# samosa = 300
# kachori = 100
# fafda = 100
# jalebi = 200
```

```python
list_size = int(input("Enter Product List Size "))
product_list = []
price = []
for i in range(list_size):
    i = input('Enter Product Name - ')
    product_list.append(i)

for i in range(list_size):
    i = int(input('Enter Price  - '))
    price.append(i)

total = sum(price)
for i in range(list_size):
        print(i+1,'. ',product_list[i],' -> ',price[i])
        
answer = input('Do You Want to add GST [Y/N]')
if answer == 'y' or answer == 'Y':
    gst = int(input('Enter GST % '))
    gst_amount = total*gst/100
    final_total = total + gst_amount
    print('Total = ',final_total,'/-')
    print('GST = ',gst,'%')
else:
    final_total = total
    print('Total = ',final_total,'/-')
    print('GST = ',0,'%')    
```
