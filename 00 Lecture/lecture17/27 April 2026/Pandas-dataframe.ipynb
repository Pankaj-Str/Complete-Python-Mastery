```python
# import pandas 
import pandas as pd

data = {
    'Name':['Payal','Poonam','Joy','Mrudula','Asmi','Sejal'],
    'Age':[23,34,56,10,78,56],
    'City':['Pune','Mumbai','Surat','Bhuj','Jaipur','Goa']
}
# use Data frame

new_index = [101,102,103,104,105,106]

df = pd.DataFrame(data,index=new_index)
df 
```

```python
# first 5 row

df.head() # default 5
```

```python
df.head(2) # print only 2 row
```

```python
# print last 2 rows 
df.tail(2) # default 5 
```

```python
# information 
df.info()
```

```python
# numeric summary 
df.describe()
```

```python
# columns print 
print(df['Name'])
```

```python
# print multiple columns 
print(df[['Name','Age']])
```

```python
# print rows iloc[index]
print(df.iloc[3])

# print multiple rows 
print(df.iloc[1:3])
```

```python
# Flitering 
print(df[df['Age'] > 34])
```

```python
# add new column
df['Salary'] = [50000,45000,67000,49000,89000,78000]
df
```

```python
# convert into excel

df.to_csv('Students.csv')
```
