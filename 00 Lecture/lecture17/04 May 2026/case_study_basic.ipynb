```python
import pandas as pd
# read csv
df = pd.read_csv("sales_data.csv")
# top 5 row print 
df.head()
```

```python
# get info
df.info()
```

```python
# check datatype 
df.dtypes
```

```python
# complete summary 
df.describe()
```

```python
# check null values 
df.isnull().sum()
```

```python

# check column name - >
print(df.columns.to_list())

# groupby()
# total sales by city 

city_total = df.groupby('city')['sales'].sum()
print(city_total)
```

```python
# average Quantity by category
avg_qty = df.groupby('category')['quantity'].mean().round(2)
avg_qty

# Q1. average Quantity by city 
# Q2. total sales by category
```

```python
# multiple statistics -> agg()
result = df.groupby('city').agg(
   {
       'sales' :['sum','mean','max'],
       'quantity': 'sum' 
   } 
).round(2)

print(result)
```

```python
# csv file 
result.to_csv('result.csv')
```

```python
# groupby two column (city + category) 
two_group = df.groupby(['city','category'])['sales'].sum().reset_index()
print(two_group)
```

```python
# count number of order per city
order_count = df.groupby('city').size().reset_index(name ="Number_of_order")
order_count
```
