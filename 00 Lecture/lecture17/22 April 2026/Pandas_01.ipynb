```python
import pandas as pd

pd.__version__
```

```python
data = ['java','go','ML','DL','JSP','Ruby','julia']
# print(data)
size = len(data)
for i in range(size):
    print(f'{i} = {data[i]}')
```

```python
s = pd.Series(data)
print(s)
```

```python
# create a new index 
indexx = [101,102,103,104,105,106,107]

s = pd.Series(data,index=indexx)
s
```

```python
# data - dictionary 
s_dict = {
    'a':1000,
    'b':9000,
    'c':8700,
    'd':8766,
    'e':4000
}

s = pd.Series(s_dict)
s
```

```python
# access data with index 
# iloc()
data = ['java','go','ML','DL','JSP','Ruby','julia']
s = pd.Series(data)
print(s)
print(s[4]) # access data from list with series
indexx = [101,102,103,104,105,106,107]

s = pd.Series(data,index=indexx)
print(s)
print(s[103])

print(s.iloc[3]) # call from index
```
