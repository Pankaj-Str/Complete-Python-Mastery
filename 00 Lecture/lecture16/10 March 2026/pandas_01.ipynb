```python
# pd.Series()
data = ["java","c++","python","go","jsp","ruby","r"]

for i in range(len(data)):
    print(i,data[i])
```

```python
#using pandas Series()
s = pd.Series(data)
print(s)
```

```python
# using index 
indexx = [101,102,103,104,105,106,107]
s = pd.Series(data,index=indexx)
print(s)
```
