```python
# empty set
data = set()
print(type(data))
```

```python
data = {12,12,1,1,2,3,4,5,4,4,5,3,2,2,2,2,2,2}
print(data)
```

```python
data = [12,12,1,1,2,3,4,5,4,4,5,3,2,2,2,2,2,2]
data_set = set(data)
print(data_set)
```

```python
# add()
data_set.add(999)
print(data_set)
```

```python
# update()

data_set.update([222,22,45])
print(data_set)
```

```python
data = list(data_set)
print(data)
```

```python
print(data[5:8])
print(data[8])
```

```python
data = {12,13,14,15,12,1,2}
data.discard(14)
print(data)
```

```python
data = {12,13,14,15,12,1,2}
data.remove(12)
print(data)
```

```python
# pop()
data.pop()
print(data)
```

```python
# CLEAR()
data.clear()
```
