```python
find_data = input('enter your word - ')
data = "Python is best"

if find_data in data:
    print(f"{find_data} :- is find ... ")
else:
    print(f"{find_data} : not find ....")
```

```python
import re

data = "python is best coding is is best think best python not R and R"

find_text = r'best'
# match()
match = re.search(find_text,data)

if match:
    print('Data find ...',match.group())
else:
    print('Data not found...')
```

```python
# findall()

find_data = re.findall(find_text,data)
print(find_data)
print(len(find_data))
```

```python
text = 'A number is a mathematical 1 2 5 6 7 object used to count, measure, and label. The most basic examples are the natural numbers: 1, 2, 3, 4, 5, and so 11 12 forth'

number = '\d+'

match = re.findall(number,text)
print(match)
```
