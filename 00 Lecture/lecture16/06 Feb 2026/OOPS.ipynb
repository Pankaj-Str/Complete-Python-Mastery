```python
class city:
    def city_info():
        print("Mumbai")
```

```python
class credit_card:
    def card_info():
        print("87654345678765")
```

```python
class HDFCBank(city,credit_card):
    def amount_info():
        print("13000/-")
```

```python
class Client(HDFCBank):
    def Client_info():
        print("Joy...")
```

```python
bank_client = Client
bank_client.amount_info()
bank_client.Client_info()
bank_client.city_info()
bank_client.card_info()
```
