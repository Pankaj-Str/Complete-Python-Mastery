```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier , plot_tree , export_text
from sklearn.metrics import accuracy_score, classification_report , confusion_matrix
import warnings
warnings.filterwarnings('ignore')
plt.style.use('default')
%matplotlib inline
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
print(df.shape)
# print columns 
print(df.columns.to_list())
df.isnull().sum()
```

```python
# target 
df['target'].value_counts()
df['target'].value_counts(normalize=True)*100
```

```python
# summary 
df.describe()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr()['target'].sort_values(ascending=False).to_frame(),annot=True,cmap='coolwarm')
plt.show()
```

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
# X.shape
print(X.columns.tolist())
```

```python
# train test split
# 25% Test
# 75% train data
X_train , X_test , y_tarin , y_test = train_test_split(
    X,y,
    test_size=0.25,
    random_state=42  
)

print(X_train.shape)
print(X_test.shape)
```

```python
# Decision Tree Model 
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_tarin)
```

```python
# make prediction 
y_pred = model.predict(X_test)
print("predicted = ",y_pred[:10])
print("Actual = ",y_test.values[:10])
```

```python
# accuracy 
acc = accuracy_score(y_test,y_pred)
print("accuracy = ",acc)

# classification report 
print("classification report : - ",classification_report(y_test,y_pred))

# confusion matrix 
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("actual")
plt.show()
```

```python
plt.figure(figsize=(50,50))
plot_tree(model,feature_names=X.columns,class_names=['No Disease','Disease'],filled=True,fontsize=20)
plt.show()
```

```python
tree_rules = export_text(model,feature_names=list(X.columns))
print(tree_rules)
```
