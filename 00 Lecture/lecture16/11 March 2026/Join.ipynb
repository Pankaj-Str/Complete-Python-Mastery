```python
import pandas as pd


students = {
    'studentID':[101,102,103,104,108],
    'name':['Rupali','Joy','Bob','Charlie','Diana'],
    'age':[23,45,67,89,12],
}

result = {
    'studentID':[101,102,103,107,109],
    'subject':['maths','java','c++','go','ruby'],
    'score':[56,78,56,90,20]
}
```

```python
students = pd.DataFrame(students)
students
```

```python
result = pd.DataFrame(result)
result
```

```python
# pd.merge()
left_merge = pd.merge(students,result,on='studentID',how='left')
left_merge
```

```python
right_merge = pd.merge(students,result,on='studentID',how='right')
right_merge
```

```python
outer_merge = pd.merge(students,result,on='studentID',how='outer')
outer_merge
```
