```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Training Data
X = np.array([
    [12,450],  # C1 Budget
    [35,2200], # C2 Premium
    [8,300],   # C3 Budget
    [20,1800], # C4 Premium
    [15,600],  # C5 Budget
    [40,2500]  # C6 Premium
])

y =np.array(['Budget','Premium','Budget','Premium','Budget','Premium'])

# new customer
new_customer = np.array([[22,3200]])
# scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
new_customer_scaled = scaler.transform(new_customer)
#print(X_scaled)
# KNN model 
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_scaled,y)
# prediction 
prediction = knn.predict(new_customer_scaled)
print("new customer to = ",prediction[0])
```

```python
from sklearn.cluster import KMeans
# Training Data
customer = np.array([
    [12,450],  # C1 
    [35,2200], # C2 
    [8,300],   # C3 
    [20,1800], # C4 
    [15,600],  # C5 
    [40,2500]  # C6 
])
KMeans_c = KMeans(
    n_clusters=2,
    random_state=0
)
KMeans_c.fit_transform(customer)
# get group 
groups = KMeans_c.labels_
centers = KMeans_c.cluster_centers_
print(groups)
print(centers)
plt.scatter(customer[:,0],customer[:,1] , c=groups,s=50, cmap='viridis',label='Customers')
plt.scatter(centers[:,0],centers[:,1] , c='red',s=200,label='center',marker='X')
plt.xlabel("Month")
plt.ylabel("rupee")
plt.legend()
plt.show()
```
