```python
class HDFC:
    def __init__(a):
        amount = 5000
        a.amount = amount
        print("HDFC BANK")
        print("Joy Jain")
        print("Borivali West ----")
        print(f'Amount = {amount}')
        
    def credit_amount(a,c_amount):
        a.c_amount = c_amount
        print(f"Credit Amount : {c_amount}/-")    
    
    def get_amount(a):
        final_amount = a.amount + a.c_amount
        print(f"Your Amount is {final_amount}")                 
```

```python
my_bank = HDFC()
my_bank.credit_amount(450)
my_bank.get_amount()

my_bank.credit_amount(41450)
my_bank.get_amount()
```
