```python
print(45 < 56)
print(45 > 56)

print(45 == 56)

print(45 != 56)

print(45 <= 56)

print(45 <= 45)

print(45 >= 56)

print(45 >= 45)
```

```python
# and 

print(23 < 67 and 80 != 80)

# or 

print(23 < 67 or 80 != 80)

# not

print(not 23 == 23)

print(not 23 != 23)
```

```python
# is ==

data1 = "python"
data2 = "java"
print(data1 is data2)

# is not  !=

print(data1 is not data2)
```

```python
# in 
data = "Python"
print('t' in data)

# not in
data = "Python"
print('t' not in data)
```
