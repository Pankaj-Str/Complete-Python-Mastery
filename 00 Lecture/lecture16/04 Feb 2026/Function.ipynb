```python
# example - 
def bill(price,gst):
    print('----------bill-----------')
    print(f'price = {price}₹')
    print(f'gst = {gst}%')
    gst_amount = price*gst/100
    print(f'GST amount = {gst_amount}₹')
    total = price+gst_amount
    print(f'Total = {total}₹')
    print('------------------------------')
```

```python
bill(1200,18)
bill(3500,9)
bill(4500,2)
bill(85000,21)
```

```python
# arg 
# example - 
def bill(price=1200,gst=18,qty=2):
    print('----------bill-----------')
    print(f'price = {price}₹')
    print(f'QTY = {qty}₹')
    print(f'price + QTY = {price*qty}₹')
    print(f'gst = {gst}%')
    gst_amount = price*gst/100
    print(f'GST amount = {gst_amount}₹')
    total = price+gst_amount
    print(f'Total = {total}₹')
    print('------------------------------')
```

```python
bill(1200,18)
bill(1200,18,10)
bill()
```
