```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,15,10,5,10] 
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.show()
```

```python
import numpy as np

x = np.linspace(0,10,100) # 100 number , 0 start 10 end
y = np.sin(x) # new data
z = np.cos(x)

plt.plot(x,y,z)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.legend(['data','cos'])
plt.show()
```

```python
# bar plot
categories = ["Apple","Watermelon","Orange","Strawberry"]
# categories = [1,52,85,56]
sales_23 = [50,40,20,60]
sales_24 = [45,23,19,80]

plt.bar(categories,sales_23,color=["red","green","pink","purple"],label="sales_data")
plt.title("Sales_of_Fruits")
#plt.xlabel("Fruits")
#plt.ylabel("Sales")
#plt.legend()
plt.xlabel("sales")
plt.ylabel("product")
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

```python
# bar plot
categories = ["Apple","Watermelon","Orange","Strawberry"]

sales_23 = [50,40,20,90]
sales_24 = [45,23,19,60]
sales_26 = [10,15,20,23]

x = np.arange(len(categories))

plt.bar(x-0.1,sales_23,label="sales 2023",width=0.4)
plt.bar(x-0.2,sales_24,label="sales 2024",width=0.4)
plt.bar(x-0.3,sales_26,label="sales 2026",width=0.4)
plt.title("Sales_of_Fruits")
plt.xticks(x,categories)
plt.xlabel("sales")
plt.ylabel("product")
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```
