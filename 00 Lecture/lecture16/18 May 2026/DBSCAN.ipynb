```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
points = np.array(
    [
        [1,1],
        [1.5,1.5],
        [2,1],
        [5,5],
        [5.5,5.5],
        [6,5],
        [5.5,6],
        [6.5,6.5],
        [10,10],
        [2.5,2.5],
        [2,2],
        [3,3]
    ]
)

# apply DBSCAN
db = DBSCAN(eps=1.9,min_samples=4).fit(points)
# get cluster labels (-1 noise)
labels = db.labels_

# visualize the clusters
plt.figure(figsize=(8,6))
unique_labels = set(labels)
colors = ['r','b','g','y'] +['k'] # color for clusters and noise

for k , col in zip(unique_labels,colors):
    if k == -1:
        col = 'k'
    class_member_mask = (labels == k)    
    xy = points[class_member_mask]
    plt.plot(xy[:, 0],xy[:, 1], 'o',
             markerfacecolor=col,
             markeredgecolor='k',
             markersize = 10,
             label = f"cluster {k}" if k != -1 else 'Noise')
    
plt.title("DBSCAN")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
