```python
try:
    height = int(input('Enter Your Height '))
    age = int(input('Enter Your Age '))
    result = height/age
    print(f"Your BMI = {result}")
except ValueError:
    print('Enter Only Number ') 
except ZeroDivisionError:
    print('Do not enter zero')       
```
