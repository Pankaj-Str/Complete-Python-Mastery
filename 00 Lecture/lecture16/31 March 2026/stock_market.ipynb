```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
stock = yf.download("RELIANCE.NS", start="2023-01-01",end="2024-01-01")
stock.head()
```

```python
# stock.to_csv("stock.csv")
```

```python
plt.Figure()
plt.plot(stock['Close'])
plt.title("RELIANCE stock plot")
plt.xlabel("Date")
plt.ylabel("Price INR")
plt.show()
```

```python
# moving avg
stock['MA50'] = stock['Close'].rolling(window=50).mean()

plt.plot(stock['Close'],label='Close Price')
plt.plot(stock['MA50'],label='50 Day')
plt.legend()
plt.show()
```

```python
stock.head()
```

```python
# bar chart
plt.bar(stock.index,stock['Volume']['RELIANCE.NS'])
plt.title("Volume stock")
plt.xlabel("Date")
plt.ylabel("volume")
plt.show()
```

```python
# candlestick chart
# %pip install mplfinance
```

```python
import mplfinance as mpf
# flatten col
stock.columns = stock.columns.droplevel(1)
# convert into float
stock = stock.astype(float)

# plot

mpf.plot(stock,type='candle',volume=True,title="chart")
```

```python
# multiple Stock comparison
stock = yf.download(["RELIANCE.NS","TCS.NS"],start="2023-01-01",end="2024-01-01")

plt.plot(stock['Close']['RELIANCE.NS'],label="RELIANCE")
plt.plot(stock['Close']['TCS.NS'],label="TCS")
plt.legend()
plt.title('Reliance vs tcs')
plt.show()
```
