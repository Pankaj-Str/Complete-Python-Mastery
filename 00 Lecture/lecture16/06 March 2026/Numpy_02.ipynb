```python
import numpy as np
import statistics as st
```

```python
data = [56,34,35,35,89,100,45,35,10,90,67]
# mean() 
mean_value = np.mean(data)
print(mean_value)
# median()
median = np.median(data)
print(median)
# mode()
mode = st.mode(data)
print(mode)
```

```python
# range()
data = [56,40,23,25,26,67]

print(min(data))
print(max(data))

range_value = max(data) - min(data)
print(range_value)
```

```python
# variance 
v = np.var(data)
print(v)
```

```python
st_var = np.std(data)
print(st_var)
```
