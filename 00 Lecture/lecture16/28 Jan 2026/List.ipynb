```python
data = ['java', 'c++', 'python', 'r', 'jsp', 'rubby']

# update

data[3] = "go"
print(data)
 
```

```python
# append()
data.append("julia")
print(data)
```

```python
# extend()
data.extend(["JSF","Spring","API"])
print(data)
```

```python
# insert()
data.insert(3,"CWPC")
print(data)
```

```python
# remove()
data.remove("java")
print(data)
```

```python
# pop()
data.pop()
print(data)
```
