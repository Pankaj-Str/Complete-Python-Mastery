```python
import pandas as pd
# melt() 

# example :- 
wide_df = pd.DataFrame(
    {
        'store':['North','South','East'],
        '2021':[1200,800,900],
        '2022':[1500,900,1100],
        '2023':[1400,850,1000]    
    }
)

wide_df
```

store | year |  sales
North | 2021 |  1200

```python
# unpivot pd.melt()
long_df = pd.melt(
    wide_df, # data
    id_vars=['store'], # columns
    value_vars=['2021','2022'], # column to melt
    var_name='year', # new col
    value_name='sales' # new col    
)

long_df
```

```python
long_df.to_csv("data.csv")
```
