```python
# import lib 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# set theme - white', 'dark', 'whitegrid', 'darkgrid', 'ticks'
sns.set_theme(style='whitegrid', palette="pastel")
plt.rcParams['figure.figsize'] = (10,6) # default
```

```python
# https://github.com/mwaskom/seaborn-data
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
# simple scatter plot two variable 
# total bill
# tips
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)
plt.title("Total bill vs tips")
plt.show()
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex", # color by gender
    style="smoker", # differents
    size="size", # size of dot
    palette="deep"
)
plt.title("Total bill vs tips")
plt.show()
```

```python
# line plot 
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(
    data=flights,
    x="year",
    y="passengers",
    hue="month"
)
plt.title("Airline data")
plt.show()
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    col="time",
    row="day")
plt.show()
```

```python
# histogram
sns.histplot(
    data=tips,
    x="total_bill",
    kde=True,
    bins=50
    )
plt.title("distribution of total bill")
plt.show()
```

```python
# category
sns.displot(
    data=tips,
    x="total_bill",
    kde=True,
    bins=30,
    hue="sex"
)
plt.title("distribution of total bill")
plt.show()
```

```python
# categorical plot
sns.barplot(data=tips,x="day",y="total_bill",hue="sex",estimator=np.mean)
plt.title("Average total bill by Day")
plt.show()
```

```python
# count plot 
sns.countplot(data=tips,x="day",hue="sex")
plt.title("Count customers per day")
plt.show()
```

```python
# box plot 
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex")
plt.title("Average total bill by Day")
plt.show()
```

```python
# box plot 
sns.violinplot(data=tips,x="day",y="total_bill",hue="sex" , split=True)
plt.title("Average total bill by Day")
plt.show()
```

```python
# Strip
sns.stripplot(data=tips,x="day",y="total_bill",hue="sex",jitter=True)
plt.title("stripplot total bill by Day")
plt.show()
```

```python
# swar
sns.swarmplot(data=tips,x="day",y="total_bill",hue="sex")
plt.title("stripplot total bill by Day")
plt.show()
```
