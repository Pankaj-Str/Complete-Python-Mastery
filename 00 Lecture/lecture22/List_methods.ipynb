```python
# List methods()
data = [600,200,300,400,400]
print(data)
# check datatype
print(type(data))
```

```python
# add new data into list (last)
data.append(401)
print(data)
```

```python
# add data from index 
data.insert(3,506)
print(data)
```

```python
# add multipal record 
data.extend([20,21,23])
print(data)
```

```python
# remove data 
# remove()
data.remove(506)
print(data)
```

```python
# pop()
data.pop(2)
print(data)
```

```python
data.pop()
print(data)
```

```python
# index()
print(data.index(401))
```

```python
data = [600,200,400,401,20,20,30,20]
# count()
print(data.count(20))
```

```python
# sort()
data.sort()
print(data)
```

```python
# reverse()
data.reverse()
print(data)
```

```python
# copy()
new_list = data.copy()
print(new_list)
```

```python
# sum()
print(sum(data))
# min()
print(min(data))
# max()
print(max(data))
```

```python
# clear()
data.clear()
data
```
