```python
data = {12,13,12,14,12,15,15,13,12,1,1,1,2,3,2,3}
print(data)
```

```python
# add() single item add
data.add(777)
print(data)
```

```python
# update() add multiple item into set

data.update([22,33,45])
print(data)
```

```python
# remove()	Removes item	Yes (raises error)
data.remove(13)
```

```python
# discard()	Removes item (safer)
data.discard(777)
print(data)
```

```python
# pop()	Removes and returns one random item	No (works on empty too)
data.pop()
print(data)
```

```python
# clear()	Removes all items	No
data.clear()
print(data)
```
