```python
# Dictionary
# data = {key:Value,Key:Value}

# example 
data = {
    'Name':'Nishant',
    'Age':23,
    'Height':4.3
    }
print(data)
```

```python
# access data from Dictionary

print(data['Name'])
print(data['Age'])
```

```python
# update data 
data['Age'] = 29
print(data)
```

```python
# add new data 
data['City'] = 'Mumbai'
print(data)
```

```python
info = {
    'Name':['Rohan','Naresh','Ganesh','Nishant'],
    'Age':[23,24,26,28],
    'City':['Suart','Mumbai','Pune','Bhuj']
}
print(info)
```

```python
data = {
        "Name": "Bhawani Shankar",
        "Description": "null",
        "BranchType": "Sub Post Office",
        "DeliveryStatus": "Delivery",
        "Circle": "Maharashtra",
        "District": "Mumbai",
        "Division": "Mumbai  West",
        "Region": "Mumbai",
        "Block": "Mumbai",
        "State": "Maharashtra",
        "Country": "India",
        "Pincode": "400028"
      }

for i in data:
    print(i,' = ',data[i])
```

```python
# keys(): Returns a list of all keys in the dictionary.
print(data.keys())
# values(): Returns a list of all values in the dictionary.
print(data.values())
# items(): Returns a list of key-value pairs as tuples.
print(data.items())
# get(key, default): Returns the value for a key if it exists, else returns a default value.
print(data.get('Name'))
# pop(key, default): Removes and returns the value associated with the key.
print(data.pop('Name'))
```
