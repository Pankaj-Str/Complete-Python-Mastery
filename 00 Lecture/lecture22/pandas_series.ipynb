```python
# pandas Series 
data = ["java","C++","JSP","Python","Pandas"]

for i in range(len(data)):
    print(f"{i}  {data[i]}")
```

```python
s = pd.Series(data)
print(s)
```

```python
# update index
new_index = [101,102,103,104,105]
s = pd.Series(data,index=new_index)
print(s)
```
